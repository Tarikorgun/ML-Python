```python
# --- TEIL 0: Bibliotheken importieren und Setup ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb
import tensorflow as tf
import pastas as ps
import pyarrow.dataset as ds
import shap
import gc
import io

from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from scikeras.wrappers import KerasClassifier # Wird wieder für die Pipeline-Integration benötigt
from statsmodels.tsa.seasonal import STL
from tqdm import tqdm
from contextlib import redirect_stdout


# Loglevel reduzieren
ps.set_log_level("ERROR")
print("Starte Skript: Stacking-Pipeline (LGBM + Neuronales Netz)")

# Setzen von Random Seeds für Reproduzierbarkeit
np.random.seed(42)
tf.random.set_seed(42)


# --- Definition der Features (UNVERÄNDERT) ---
STATIC_CATEGORICAL_FEATURES = ["main_landuse"]
STATIC_NUMERICAL_FEATURES = ["ground_elevation_m_asl", "topographic_slope_degrees"]
DYNAMIC_NUMERICAL_FEATURES_RAW = [
    "groundwater_level_m_asl", "precipitation_gpcc_mm_year-1",
    "potential_evapotranspiration_era5_mm_year-1", "air_temperature_°C", "snow_depth_m",
    "ndvi_ratio", "lai_low_vegetation_ratio", "lai_high_vegetation_ratio",
    "withdrawal_industrial_m3_year-1", "withdrawal_domestic_m3_year-1"
]

# --- TEIL 1 & 2: DATEN LADEN UND FEATURE ENGINEERING (UNVERÄNDERT) ---
print("\n" + "="*80); print(" TEIL 1 & 2: DATEN LADEN UND FEATURE ENGINEERING"); print("="*80)
# (Der gesamte Code aus deinem Original-Skript für Teil 1 & 2 bleibt hier unverändert)
# ...
# ANNAHME: Am Ende dieses Teils existieren wie zuvor die DataFrames:
# X_all_features_raw, y_all_labels, X_meta_for_split
# Aus Übersichtlichkeitsgründen wird der Code hier nicht erneut eingefügt.
# START DER DATENVERARBEITUNG (KOPIERT)
path_attributes = "C:/MLmitPython/ML-Python/Daten/grow_attributes.parquet"
path_timeseries = "C:/MLmitPython/ML-Python/Daten/grow_timeseries.parquet"
try:
    attr_df_full = pd.read_parquet(path_attributes)
except FileNotFoundError: raise
attr_filtered_df = attr_df_full[(attr_df_full["country"] == "CZE") & (attr_df_full["length_years"] >= 30) & (attr_df_full["gap_fraction"] == 0.0) & (pd.to_datetime(attr_df_full["ending_date"]) >= pd.Timestamp("2023-12-01"))]
if attr_filtered_df.empty: raise ValueError("Keine passende Zeitreihe für CZ für ML gefunden.")
cz_ids_for_ml = attr_filtered_df["GROW_ID"].dropna().unique().tolist()
dynamic_feature_columns_to_load = ["GROW_ID", "date"] + DYNAMIC_NUMERICAL_FEATURES_RAW
try:
    timeseries_dataset_raw = ds.dataset(path_timeseries, format="parquet")
    available_columns_in_parquet = timeseries_dataset_raw.schema.names
    columns_to_actually_load = [col for col in dynamic_feature_columns_to_load if col in available_columns_in_parquet]
    timeseries_table_raw = timeseries_dataset_raw.to_table(filter=ds.field("GROW_ID").isin(cz_ids_for_ml), columns=columns_to_actually_load)
    raw_ts_df_full = timeseries_table_raw.to_pandas(self_destruct=True, split_blocks=True, zero_copy_only=False)
except FileNotFoundError: raise
raw_ts_df_full["date"] = pd.to_datetime(raw_ts_df_full["date"])
if "precipitation_gpcc_mm_year-1" in raw_ts_df_full.columns: raw_ts_df_full["precipitation_gpcc_mm_month-1"] = (raw_ts_df_full["precipitation_gpcc_mm_year-1"] / 12)
if "potential_evapotranspiration_era5_mm_year-1" in raw_ts_df_full.columns: raw_ts_df_full["potential_evapotranspiration_era5_mm_month-1"] = (raw_ts_df_full["potential_evapotranspiration_era5_mm_year-1"] / 12)
grouped_ts_data = raw_ts_df_full.set_index('date').groupby('GROW_ID'); del raw_ts_df_full; gc.collect()
def process_grow_id_data(grow_id_data):
    gwl_series_raw = grow_id_data["groundwater_level_m_asl"].dropna()
    if gwl_series_raw.empty or len(gwl_series_raw.resample('MS').mean().dropna()) < 120: return None
    gwl_series_monthly = gwl_series_raw.resample('MS').mean()
    try:
        with redirect_stdout(io.StringIO()):
            sgi_observed_monthly = ps.stats.sgi(gwl_series_monthly)
        if sgi_observed_monthly.empty or sgi_observed_monthly.isnull().all(): return None
    except Exception: return None
    sgi_trend, sgi_seasonal, sgi_resid = (pd.Series(dtype=np.float32), pd.Series(dtype=np.float32), pd.Series(dtype=np.float32))
    sgi_for_stl = sgi_observed_monthly.dropna()
    if len(sgi_for_stl) >= 25:
        try:
            stl = STL(sgi_for_stl, period=12, robust=True); res_stl = stl.fit()
            sgi_trend, sgi_seasonal, sgi_resid = (res_stl.trend.reindex(sgi_observed_monthly.index), res_stl.seasonal.reindex(sgi_observed_monthly.index), res_stl.resid.reindex(sgi_observed_monthly.index))
        except Exception: pass
    monthly_features_dict = {'sgi_label': sgi_observed_monthly, 'sgi_trend': sgi_trend, 'sgi_seasonal': sgi_seasonal, 'sgi_resid': sgi_resid}
    dynamic_map = {"precipitation_gpcc_mm_month-1": "precipitation_monthly", "potential_evapotranspiration_era5_mm_month-1": "evapotranspiration_monthly", "air_temperature_°C": "air_temperature_monthly", "snow_depth_m": "snow_depth_monthly", "ndvi_ratio": "ndvi_monthly", "lai_low_vegetation_ratio": "lai_low_veg_monthly", "lai_high_vegetation_ratio": "lai_high_veg_monthly"}
    for orig, target in dynamic_map.items(): monthly_features_dict[target] = grow_id_data[orig].dropna().resample('MS').mean().reindex(sgi_observed_monthly.index) if orig in grow_id_data.columns else pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    yearly_map = {"withdrawal_industrial_m3_year-1": "withdrawal_industrial_monthly", "withdrawal_domestic_m3_year-1": "withdrawal_domestic_monthly"}
    for orig, target in yearly_map.items():
        if orig in grow_id_data.columns:
            series_yearly = grow_id_data[orig].dropna(); series_yearly = series_yearly[~series_yearly.index.duplicated(keep='first')] if not series_yearly.index.is_unique else series_yearly; monthly_features_dict[target] = series_yearly.resample('MS').ffill().reindex(sgi_observed_monthly.index)
        else: monthly_features_dict[target] = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    return pd.DataFrame(monthly_features_dict).dropna(subset=['sgi_label'])
processed_groups = []
for grow_id, group_data in tqdm(grouped_ts_data, desc="SGI-Berechnung & Dekomposition"):
    processed_df = process_grow_id_data(group_data)
    if processed_df is not None and not processed_df.empty: processed_df['GROW_ID'] = grow_id; processed_groups.append(processed_df)
if not processed_groups: raise ValueError("Keine Daten für ML-Modell nach Verarbeitung.")
df_ml_input_dynamic = pd.concat(processed_groups); del processed_groups; gc.collect()
static_features_to_select = ["GROW_ID"] + STATIC_NUMERICAL_FEATURES + STATIC_CATEGORICAL_FEATURES
static_features_to_select = [col for col in static_features_to_select if col in attr_df_full.columns]
attr_df_ml_static = attr_df_full[attr_df_full['GROW_ID'].isin(cz_ids_for_ml)][static_features_to_select].copy().drop_duplicates(subset=['GROW_ID'], keep='first')
df_ml_input_dynamic.reset_index(inplace=True)
df_ml_input = pd.merge(df_ml_input_dynamic, attr_df_ml_static, on="GROW_ID", how="left").set_index('date')
del df_ml_input_dynamic, attr_df_ml_static, attr_df_full, attr_filtered_df; gc.collect()
df_ml_processed = df_ml_input.copy(); del df_ml_input; gc.collect()
for lag_col_base in ['sgi_label', 'sgi_trend', 'sgi_seasonal', 'sgi_resid']:
    if lag_col_base in df_ml_processed.columns:
        for lag_val in [1, 3, 6, 12]: df_ml_processed[f'{lag_col_base}_lag_{lag_val}M'] = df_ml_processed.groupby('GROW_ID')[lag_col_base].shift(lag_val)
df_ml_processed['month_num'] = df_ml_processed.index.month.astype(np.int8)
df_ml_processed['year'] = df_ml_processed.index.year.astype(np.int32)
VORHERSAGE_HORIZONT_MONATE = 6
sgi_target_col_continuous = f'sgi_target_{VORHERSAGE_HORIZONT_MONATE}M_future'
df_ml_processed[sgi_target_col_continuous] = df_ml_processed.groupby('GROW_ID')['sgi_label'].shift(-VORHERSAGE_HORIZONT_MONATE)
def sgi_to_drought_class(z_score):
    if pd.isna(z_score): return None
    if z_score > -0.84: return 'Normal'
    if -1.50 <= z_score <= -0.84: return 'Mild_Moderate'
    if -2.00 <= z_score <= -1.51: return 'Severe'
    return 'Extreme'
df_ml_processed['drought_class_str'] = df_ml_processed[sgi_target_col_continuous].apply(sgi_to_drought_class)
class_mapping = {'Normal': 0, 'Mild_Moderate': 1, 'Severe': 2, 'Extreme': 3}
df_ml_processed['drought_class_target'] = df_ml_processed['drought_class_str'].map(class_mapping).astype('Int64')
label_column_ml = 'drought_class_target'
categorical_cols_to_encode = [col for col in STATIC_CATEGORICAL_FEATURES if col in df_ml_processed.columns] + ['month_num']
dummy_encoded_cols_list = []
for col_to_encode in categorical_cols_to_encode:
    dummies = pd.get_dummies(df_ml_processed[col_to_encode], prefix=col_to_encode, dummy_na=False, dtype=np.int8)
    df_ml_processed = pd.concat([df_ml_processed, dummies], axis=1); df_ml_processed.drop(columns=[col_to_encode], inplace=True)
    dummy_encoded_cols_list.extend(dummies.columns.tolist())
numerical_static_cols_for_imputation = [col for col in STATIC_NUMERICAL_FEATURES if col in df_ml_processed.columns]
if numerical_static_cols_for_imputation:
    df_ml_processed[numerical_static_cols_for_imputation] = SimpleImputer(strategy='median').fit_transform(df_ml_processed[numerical_static_cols_for_imputation])
base_features_ml = ["precipitation_monthly", "evapotranspiration_monthly", "air_temperature_monthly", "snow_depth_monthly", "ndvi_monthly", "lai_low_veg_monthly", "lai_high_veg_monthly", "withdrawal_industrial_monthly", "withdrawal_domestic_monthly"]
lag_features_ml = [col for col in df_ml_processed.columns if '_lag_' in col]
current_sgi_features = ['sgi_label', 'sgi_trend', 'sgi_seasonal', 'sgi_resid']
features_columns_ml = list(set([col for col in base_features_ml if col in df_ml_processed.columns] + lag_features_ml + current_sgi_features + numerical_static_cols_for_imputation + dummy_encoded_cols_list))
columns_to_check_for_dropna_smart = [label_column_ml, 'sgi_label_lag_1M', 'precipitation_monthly']
df_ml_processed.dropna(subset=[col for col in columns_to_check_for_dropna_smart if col in df_ml_processed.columns], inplace=True)
df_ml_processed[label_column_ml] = df_ml_processed[label_column_ml].astype(int)
X_all_features_raw = df_ml_processed[features_columns_ml].copy()
y_all_labels = df_ml_processed[label_column_ml].copy()
X_meta_for_split = df_ml_processed[['GROW_ID', 'year']].copy()
del df_ml_processed; gc.collect()
print("\n[SUCCESS] Teil 1 & 2 abgeschlossen.")
# ENDE DER DATENVERARBEITUNG


# --- TEIL 3: TRAINING MEHRERER MODELLE & FINALES HYBRID ---
print("\n" + "="*80); print(" TEIL 3: TRAINING MEHRERER MODELLE & FINALES HYBRID"); print("="*80)

# --- Zeitbasierter Split (UNVERÄNDERT) ---
temp_year_series = X_meta_for_split['year']
unique_years_ml = np.sort(temp_year_series.unique())
split_year_1 = int(np.percentile(unique_years_ml, 80))
split_year_2 = int(np.percentile(unique_years_ml, 90))
# Wir brauchen wieder alle drei Sets, um die Einzelmodelle und das Stacking-Modell zu vergleichen
train_mask = temp_year_series < split_year_1
val_mask = (temp_year_series >= split_year_1) & (temp_year_series < split_year_2)
test_mask = temp_year_series >= split_year_2
X_train_raw, X_val_raw, X_test_raw = X_all_features_raw[train_mask], X_all_features_raw[val_mask], X_all_features_raw[test_mask]
y_train, y_val, y_test = y_all_labels[train_mask], y_all_labels[val_mask], y_all_labels[test_mask]
# Für Stacking nutzen wir das kombinierte Trainings- und Validierungsset
X_train_stack = X_all_features_raw[train_val_mask]
y_train_stack = y_all_labels[train_val_mask]
print("Daten erfolgreich aufgeteilt.")


# ############################################################################
# ### 3A: TRAINING & EVALUATION DES REINEN LGBM-MODELLS ###
# ############################################################################
print("\n" + "-"*40); print(" 3A: Training des reinen LGBM-Modells "); print("-"*40)

lgbm_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('classifier', lgb.LGBMClassifier(class_weight='balanced', random_state=42, n_estimators=300, num_leaves=70, learning_rate=0.05))
])
lgbm_pipeline.fit(X_train_raw, y_train)
print("LGBM-Modell trainiert.")
print("\nEVALUATION LGBM AUF TEST-DATEN")
y_pred_lgbm = lgbm_pipeline.predict(X_test_raw)
print(classification_report(y_test, y_pred_lgbm, target_names=class_mapping.keys()))


# ############################################################################
# ### 3B: TRAINING & EVALUATION DES STACKING-MODELLS ###
# ############################################################################
print("\n" + "-"*40); print(" 3B: Training des Stacking-Modells "); print("-"*40)

# Hier kommt der komplette Stacking-Code aus dem letzten Skript rein
# (Definition der Pipelines, des Meta-Classifiers und des StackingClassifiers)
def create_nn_for_stacking(meta):
    n_features_in_ = meta["n_features_in_"]
    n_classes_ = meta["n_classes_"]
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(n_features_in_,)),
        tf.keras.layers.Dense(80, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        tf.keras.layers.Dropout(0.4),
        tf.keras.layers.Dense(40, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        tf.keras.layers.Dropout(0.4),
        tf.keras.layers.Dense(n_classes_, activation='softmax')
    ])
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.005)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

nn_classifier = KerasClassifier(model=create_nn_for_stacking, epochs=50, batch_size=256, class_weight='balanced', verbose=0)
nn_pipeline = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean')),('scaler', StandardScaler()),('classifier', nn_classifier)])
meta_classifier = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
estimators = [('lgbm', lgbm_pipeline), ('neural_net', nn_pipeline)]

stacking_model = StackingClassifier(estimators=estimators, final_estimator=meta_classifier, cv=5, stack_method='predict_proba', n_jobs=1, verbose=2)
stacking_model.fit(X_train_stack, y_train_stack)
print("Stacking-Modell trainiert.")
print("\nEVALUATION STACKING-MODELL AUF TEST-DATEN")
y_pred_stacking = stacking_model.predict(X_test)
print(classification_report(y_test, y_pred_stacking, target_names=class_mapping.keys()))


# ############################################################################
# ### 3C: ERSTELLUNG UND EVALUATION DES FINALEN HYBRID-MODELLS ###
# ############################################################################
print("\n" + "-"*40); print(" 3C: Erstellung und Evaluation des finalen Hybrid-Modells "); print("-"*40)

# Hier kommt dein Code-Schnipsel hin. Er funktioniert jetzt, weil
# y_pred_lgbm und y_pred_stacking in den Schritten zuvor erstellt wurden.
severe_class_index = class_mapping['Severe']
extreme_class_index = class_mapping['Extreme']

stacking_predicts_crisis = (y_pred_stacking == severe_class_index) | (y_pred_stacking == extreme_class_index)

final_hybrid_prediction = np.where(
    stacking_predicts_crisis,
    y_pred_stacking,
    y_pred_lgbm
)

print("\nEVALUATION DES FINALEN HYBRID-MODELLS (LGBM + STACKING)")
print(classification_report(y_test, final_hybrid_prediction, target_names=class_mapping.keys()))

print("\nFinale Konfusionsmatrix des Hybrid-Modells:")
cm = confusion_matrix(y_test, final_hybrid_prediction, labels=range(len(class_mapping)))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_mapping.keys())
fig, ax = plt.subplots(figsize=(10, 8)); disp.plot(cmap=plt.cm.Blues, ax=ax, xticks_rotation='vertical');
ax.set_title('Klassifikationsmatrix für finales Hybrid-Modell'); plt.tight_layout(); plt.show()
```

```python
# --- TEIL 0: Bibliotheken importieren und Setup ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pastas as ps
import pyarrow.dataset as ds
import tensorflow as tf # Spezifisch für Neuronale Netze (Keras/TensorFlow)
from sklearn.utils import class_weight # Für die Handhabung ungleicher Klassenverteilungen
import shap # Für die Modell-Interpretation (SHapley Additive exPlanations)
from sklearn.preprocessing import StandardScaler # Zur Skalierung numerischer Features
from sklearn.impute import SimpleImputer # Zum Füllen von fehlenden Werten
import gc # Für die manuelle Speicherbereinigung
from statsmodels.tsa.seasonal import STL # Für die saisonale Zerlegung von Zeitreihen
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay # Zur Evaluation des Modells
from tqdm import tqdm # Für schöne Fortschrittsbalken bei langen Schleifen
from contextlib import redirect_stdout # Um störende Ausgaben von Bibliotheken zu unterdrücken
import io # Für die Umleitung von Ausgaben

# --- Konfiguration ---
# Loglevel von 'pastas' reduzieren, um die Konsole sauber zu halten
ps.set_log_level("ERROR")
print("Starte Skript: Klassifikations-Pipeline mit NEURONALEM NETZ")

# Setzen von Random Seeds für Reproduzierbarkeit
np.random.seed(42)
tf.random.set_seed(42)


# --- Definition der Features (UNVERÄNDERT) ---
# Hier werden die Spaltennamen für verschiedene Feature-Typen zentral definiert.
STATIC_CATEGORICAL_FEATURES = ["main_landuse"]
STATIC_NUMERICAL_FEATURES = ["ground_elevation_m_asl", "topographic_slope_degrees"]
DYNAMIC_NUMERICAL_FEATURES_RAW = [
    "groundwater_level_m_asl", "precipitation_gpcc_mm_year-1",
    "potential_evapotranspiration_era5_mm_year-1", "air_temperature_°C", "snow_depth_m",
    "ndvi_ratio", "lai_low_vegetation_ratio", "lai_high_vegetation_ratio",
    "withdrawal_industrial_m3_year-1", "withdrawal_domestic_m3_year-1"
]

# --- TEIL 1 & 2: DATEN LADEN UND FEATURE ENGINEERING ---
print("\n" + "="*80); print(" TEIL 1 & 2: DATEN LADEN UND FEATURE ENGINEERING"); print("="*80)

# --- Schritt 1.1: Statische Attribute laden und filtern ---
print("1.1 Lade statische Attribute und filtere relevante Stationen...")
path_attributes = "C:/MLmitPython/ML-Python/Daten/grow_attributes.parquet"
path_timeseries = "C:/MLmitPython/ML-Python/Daten/grow_timeseries.parquet"
try:
    attr_df_full = pd.read_parquet(path_attributes)
except FileNotFoundError: raise

# Filtere die Attribute nach den gewünschten Kriterien (Tschechien, lange Zeitreihen, keine Lücken etc.)
attr_filtered_df = attr_df_full[(attr_df_full["country"] == "CZE") & (attr_df_full["length_years"] >= 30) & (attr_df_full["gap_fraction"] == 0.0) & (pd.to_datetime(attr_df_full["ending_date"]) >= pd.Timestamp("2023-12-01"))]
if attr_filtered_df.empty: raise ValueError("Keine passende Zeitreihe für CZ für ML gefunden.")
cz_ids_for_ml = attr_filtered_df["GROW_ID"].dropna().unique().tolist()
print(f"--> {len(cz_ids_for_ml)} relevante Stationen für die Analyse gefunden.")

# --- Schritt 1.2: Dynamische Zeitreihen laden ---
print("\n1.2 Lade dynamische Zeitreihen für die ausgewählten Stationen...")
dynamic_feature_columns_to_load = ["GROW_ID", "date"] + DYNAMIC_NUMERICAL_FEATURES_RAW
try:
    # PyArrow wird für schnelles, speichereffizientes Laden der großen Parquet-Datei verwendet
    timeseries_dataset_raw = ds.dataset(path_timeseries, format="parquet")
    available_columns_in_parquet = timeseries_dataset_raw.schema.names
    columns_to_actually_load = [col for col in dynamic_feature_columns_to_load if col in available_columns_in_parquet]
    timeseries_table_raw = timeseries_dataset_raw.to_table(filter=ds.field("GROW_ID").isin(cz_ids_for_ml), columns=columns_to_actually_load)
    raw_ts_df_full = timeseries_table_raw.to_pandas(self_destruct=True, split_blocks=True, zero_copy_only=False)
except FileNotFoundError: raise
raw_ts_df_full["date"] = pd.to_datetime(raw_ts_df_full["date"])
print("--> Zeitreihen erfolgreich geladen.")

# --- Schritt 1.3: Erste Feature-Anpassungen ---
print("\n1.3 Berechne monatliche Features aus jährlichen Werten...")
if "precipitation_gpcc_mm_year-1" in raw_ts_df_full.columns: raw_ts_df_full["precipitation_gpcc_mm_month-1"] = (raw_ts_df_full["precipitation_gpcc_mm_year-1"] / 12)
if "potential_evapotranspiration_era5_mm_year-1" in raw_ts_df_full.columns: raw_ts_df_full["potential_evapotranspiration_era5_mm_month-1"] = (raw_ts_df_full["potential_evapotranspiration_era5_mm_year-1"] / 12)

# Gruppiere die Zeitreihen nach ihrer ID für die Einzelverarbeitung
grouped_ts_data = raw_ts_df_full.set_index('date').groupby('GROW_ID'); del raw_ts_df_full; gc.collect()

# --- Schritt 1.4: Haupt-Feature-Engineering pro Station ---
print("\n1.4 Führe SGI-Berechnung und Feature-Engineering pro Station durch...")
# Definiere eine Funktion, die alle Schritte für eine einzelne Station durchführt
def process_grow_id_data(grow_id_data):
    # Resample Grundwasserstand auf monatliche Werte
    gwl_series_raw = grow_id_data["groundwater_level_m_asl"].dropna()
    if gwl_series_raw.empty or len(gwl_series_raw.resample('MS').mean().dropna()) < 120: return None # Mindestens 10 Jahre Daten benötigt
    gwl_series_monthly = gwl_series_raw.resample('MS').mean()
    
    # Berechne den Standardized Groundwater Index (SGI) mit der 'pastas'-Bibliothek
    try:
        with redirect_stdout(io.StringIO()): # Unterdrückt Konsolenausgaben von pastas
            sgi_observed_monthly = ps.stats.sgi(gwl_series_monthly)
        if sgi_observed_monthly.empty or sgi_observed_monthly.isnull().all(): return None
    except Exception: return None
    
    # Zerlege die SGI-Zeitreihe in Trend, Saison und Residuum (Rest)
    sgi_trend, sgi_seasonal, sgi_resid = (pd.Series(dtype=np.float32), pd.Series(dtype=np.float32), pd.Series(dtype=np.float32))
    sgi_for_stl = sgi_observed_monthly.dropna()
    if len(sgi_for_stl) >= 25: # STL benötigt eine minimale Anzahl an Datenpunkten
        try:
            stl = STL(sgi_for_stl, period=12, robust=True); res_stl = stl.fit()
            sgi_trend, sgi_seasonal, sgi_resid = (res_stl.trend.reindex(sgi_observed_monthly.index), res_stl.seasonal.reindex(sgi_observed_monthly.index), res_stl.resid.reindex(sgi_observed_monthly.index))
        except Exception: pass
        
    # Sammle alle berechneten und originalen Features in einem Dictionary
    monthly_features_dict = {'sgi_label': sgi_observed_monthly, 'sgi_trend': sgi_trend, 'sgi_seasonal': sgi_seasonal, 'sgi_resid': sgi_resid}
    dynamic_map = {"precipitation_gpcc_mm_month-1": "precipitation_monthly", "potential_evapotranspiration_era5_mm_month-1": "evapotranspiration_monthly", "air_temperature_°C": "air_temperature_monthly", "snow_depth_m": "snow_depth_monthly", "ndvi_ratio": "ndvi_monthly", "lai_low_vegetation_ratio": "lai_low_veg_monthly", "lai_high_vegetation_ratio": "lai_high_veg_monthly"}
    for orig, target in dynamic_map.items(): monthly_features_dict[target] = grow_id_data[orig].dropna().resample('MS').mean().reindex(sgi_observed_monthly.index) if orig in grow_id_data.columns else pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    yearly_map = {"withdrawal_industrial_m3_year-1": "withdrawal_industrial_monthly", "withdrawal_domestic_m3_year-1": "withdrawal_domestic_monthly"}
    for orig, target in yearly_map.items():
        if orig in grow_id_data.columns:
            series_yearly = grow_id_data[orig].dropna(); series_yearly = series_yearly[~series_yearly.index.duplicated(keep='first')] if not series_yearly.index.is_unique else series_yearly; monthly_features_dict[target] = series_yearly.resample('MS').ffill().reindex(sgi_observed_monthly.index)
        else: monthly_features_dict[target] = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
        
    return pd.DataFrame(monthly_features_dict).dropna(subset=['sgi_label'])

# Wende die Funktion auf alle Stations-Gruppen an und zeige einen Fortschrittsbalken
processed_groups = []
for grow_id, group_data in tqdm(grouped_ts_data, desc="SGI-Berechnung & Dekomposition"):
    processed_df = process_grow_id_data(group_data)
    if processed_df is not None and not processed_df.empty: processed_df['GROW_ID'] = grow_id; processed_groups.append(processed_df)
if not processed_groups: raise ValueError("Keine Daten für ML-Modell nach Verarbeitung.")

# --- Schritt 1.5: Finale Datenzusammenführung und Feature-Erstellung ---
print("\n1.5 Führe dynamische und statische Features zusammen...")
df_ml_input_dynamic = pd.concat(processed_groups); del processed_groups; gc.collect()
static_features_to_select = ["GROW_ID"] + STATIC_NUMERICAL_FEATURES + STATIC_CATEGORICAL_FEATURES
static_features_to_select = [col for col in static_features_to_select if col in attr_df_full.columns]
attr_df_ml_static = attr_df_full[attr_df_full['GROW_ID'].isin(cz_ids_for_ml)][static_features_to_select].copy().drop_duplicates(subset=['GROW_ID'], keep='first')
df_ml_input_dynamic.reset_index(inplace=True)
df_ml_input = pd.merge(df_ml_input_dynamic, attr_df_ml_static, on="GROW_ID", how="left").set_index('date')
del df_ml_input_dynamic, attr_df_ml_static, attr_df_full, attr_filtered_df; gc.collect()
df_ml_processed = df_ml_input.copy(); del df_ml_input; gc.collect()

print("\n1.6 Erstelle zeitversetzte (Lag) Features und Zielvariable...")
# Erstelle Lag-Features für die SGI-Komponenten (Werte von vor 1, 3, 6, 12 Monaten)
for lag_col_base in ['sgi_label', 'sgi_trend', 'sgi_seasonal', 'sgi_resid']:
    if lag_col_base in df_ml_processed.columns:
        for lag_val in [1, 3, 6, 12]: df_ml_processed[f'{lag_col_base}_lag_{lag_val}M'] = df_ml_processed.groupby('GROW_ID')[lag_col_base].shift(lag_val)

df_ml_processed['month_num'] = df_ml_processed.index.month.astype(np.int8)
df_ml_processed['year'] = df_ml_processed.index.year.astype(np.int32)

# Erstelle die Zielvariable: der SGI-Wert 6 Monate in der Zukunft
VORHERSAGE_HORIZONT_MONATE = 6
sgi_target_col_continuous = f'sgi_target_{VORHERSAGE_HORIZONT_MONATE}M_future'
df_ml_processed[sgi_target_col_continuous] = df_ml_processed.groupby('GROW_ID')['sgi_label'].shift(-VORHERSAGE_HORIZONT_MONATE)

# --- Schritt 1.7: Zielvariable in Klassen umwandeln und Features vorbereiten ---
print("\n1.7 Wandle Zielvariable in Klassen um und kodiere kategorische Features...")
def sgi_to_drought_class(z_score):
    if pd.isna(z_score): return None
    if z_score > -0.84: return 'Normal'
    if -1.50 <= z_score <= -0.84: return 'Mild_Moderate'
    if -2.00 <= z_score <= -1.51: return 'Severe'
    return 'Extreme'
df_ml_processed['drought_class_str'] = df_ml_processed[sgi_target_col_continuous].apply(sgi_to_drought_class)
class_mapping = {'Normal': 0, 'Mild_Moderate': 1, 'Severe': 2, 'Extreme': 3}
df_ml_processed['drought_class_target'] = df_ml_processed['drought_class_str'].map(class_mapping).astype('Int64')
label_column_ml = 'drought_class_target'

# Definiere die zu kodierenden Spalten dynamisch, basierend auf dem, was wirklich im DataFrame ist
print("\n1.7.1 Kodierung der kategorischen Features...")
categorical_cols_to_encode = [col for col in STATIC_CATEGORICAL_FEATURES if col in df_ml_processed.columns] + ['month_num']
print(f"--> Folgende kategorische Spalten werden kodiert: {categorical_cols_to_encode}")

# Die Schleife bleibt gleich, verarbeitet aber nur existierende Spalten
dummy_encoded_cols_list = []
for col_to_encode in categorical_cols_to_encode:
    print(f"    - Kodiere '{col_to_encode}'...")
    dummies = pd.get_dummies(df_ml_processed[col_to_encode], prefix=col_to_encode, dummy_na=False, dtype=np.int8)
    df_ml_processed = pd.concat([df_ml_processed, dummies], axis=1); df_ml_processed.drop(columns=[col_to_encode], inplace=True)
    dummy_encoded_cols_list.extend(dummies.columns.tolist())

# --- Schritt 1.8: Finale Datenbereinigung ---
print("\n1.8 Bereinige finale Datensätze...")
# Fülle verbleibende Lücken in numerischen Spalten mit dem Median
numerical_static_cols_for_imputation = [col for col in STATIC_NUMERICAL_FEATURES if col in df_ml_processed.columns]
if numerical_static_cols_for_imputation:
    df_ml_processed[numerical_static_cols_for_imputation] = SimpleImputer(strategy='median').fit_transform(df_ml_processed[numerical_static_cols_for_imputation])

# Definiere die finale Liste aller Features für das Modell
base_features_ml = ["precipitation_monthly", "evapotranspiration_monthly", "air_temperature_monthly", "snow_depth_monthly", "ndvi_monthly", "lai_low_veg_monthly", "lai_high_veg_monthly", "withdrawal_industrial_monthly", "withdrawal_domestic_monthly"]
lag_features_ml = [col for col in df_ml_processed.columns if '_lag_' in col]
current_sgi_features = ['sgi_label', 'sgi_trend', 'sgi_seasonal', 'sgi_resid']
features_columns_ml = list(set([col for col in base_features_ml if col in df_ml_processed.columns] + lag_features_ml + current_sgi_features + numerical_static_cols_for_imputation + dummy_encoded_cols_list))
# Entferne Zeilen, in denen wichtige Features oder das Label fehlen
columns_to_check_for_dropna_smart = [label_column_ml, 'sgi_label_lag_1M', 'precipitation_monthly']
df_ml_processed.dropna(subset=[col for col in columns_to_check_for_dropna_smart if col in df_ml_processed.columns], inplace=True)
df_ml_processed[label_column_ml] = df_ml_processed[label_column_ml].astype(int)

# Finale Aufteilung in X (Features) und y (Zielvariable)
X_all_features_raw = df_ml_processed[features_columns_ml].copy()
y_all_labels = df_ml_processed[label_column_ml].copy()
X_meta_for_split = df_ml_processed[['GROW_ID', 'year']].copy()
del df_ml_processed; gc.collect()
print("\n[SUCCESS] Teil 1 & 2 abgeschlossen.")


# --- TEIL 3: MODELLTRAINING UND EVALUATION ---
print("\n" + "="*80); print(" TEIL 3: MODELLTRAINING & EVALUATION (NEURONALES NETZ)"); print("="*80)

# --- Schritt 3.1: Zeitbasierter Train-Validation-Test Split ---
temp_year_series = X_meta_for_split['year']
unique_years_ml = np.sort(temp_year_series.unique())
split_year_1 = int(np.percentile(unique_years_ml, 80))
split_year_2 = int(np.percentile(unique_years_ml, 90))
print(f"--> Erstelle zeitbasierten 80/10/10 Split:")
print(f"    - Trainingsdaten: Jahre < {split_year_1} (80%)")
print(f"    - Validierungsdaten: Jahre >= {split_year_1} und < {split_year_2} (10%)")
print(f"    - Testdaten: Jahre >= {split_year_2} (10%)")
train_mask = temp_year_series < split_year_1
val_mask = (temp_year_series >= split_year_1) & (temp_year_series < split_year_2)
test_mask = temp_year_series >= split_year_2
X_train_raw, X_val_raw, X_test_raw = X_all_features_raw[train_mask], X_all_features_raw[val_mask], X_all_features_raw[test_mask]
y_train, y_val, y_test = y_all_labels[train_mask], y_all_labels[val_mask], y_all_labels[test_mask]
print(f"\nGröße der Datensätze:")
print(f"    - Training:   {X_train_raw.shape[0]} Zeilen")
print(f"    - Validation: {X_val_raw.shape[0]} Zeilen")
print(f"    - Test:       {X_test_raw.shape[0]} Zeilen")
del X_all_features_raw, y_all_labels, X_meta_for_split; gc.collect()


# --- Schritt 3.2: Finale Feature-Übersicht ---
print("\n" + "-"*80)
features_in_model = sorted(X_train_raw.columns.tolist())
print(f"--> Folgende {len(features_in_model)} Features werden für das Modell verwendet:")
print("-" * 80)
max_len = max(len(f) for f in features_in_model) + 2
num_columns = 4
for i in range(0, len(features_in_model), num_columns):
    row = [f"{feat:<{max_len}}" for feat in features_in_model[i:i+num_columns]]
    print("".join(row))
print("-" * 80)


# --- Schritt 3.3: Manuelles Preprocessing der Daten (vor dem NN-Training) ---
print("\n3.3 Führe manuelles Preprocessing durch (Imputer & Scaler)...")
# Imputer zum Füllen fehlender Werte (z.B. mit dem Mittelwert)
imputer = SimpleImputer(strategy='mean')
# Scaler zum Skalieren der Daten (z.B. auf Mittelwert 0 und Standardabweichung 1)
scaler = StandardScaler()

# Daten transformieren: zuerst Imputation, dann Skalierung
X_train_scaled = scaler.fit_transform(imputer.fit_transform(X_train_raw))
X_val_scaled = scaler.transform(imputer.transform(X_val_raw))
X_test_scaled = scaler.transform(imputer.transform(X_test_raw))
print("--> Daten erfolgreich imputiert und skaliert.")


# --- Schritt 3.4: Definition des Keras-Modells (Neuronales Netz) ---
print("\n3.4 Definiere die Architektur des Neuronalen Netzes...")
n_features = X_train_scaled.shape[1] # Anzahl der Eingabefeatures
n_classes = len(class_mapping) # Anzahl der Zielklassen

# Definition des sequentiellen Modells
model = tf.keras.Sequential([
    # Input-Layer: Definiert die Form der Eingabedaten
    tf.keras.layers.Input(shape=(n_features,)),
    # Erste Hidden-Layer: 80 Neuronen, ReLU-Aktivierung, L2-Regularisierung gegen Overfitting
    tf.keras.layers.Dense(
        80, 
        activation='relu',
        kernel_regularizer=tf.keras.regularizers.l2(0.001) 
    ),
    # Dropout-Layer: Deaktiviert zufällig 40% der Neuronen, um Overfitting zu reduzieren
    tf.keras.layers.Dropout(0.4), 
    # Zweite Hidden-Layer: 40 Neuronen, ReLU-Aktivierung, L2-Regularisierung
    tf.keras.layers.Dense(
        40, 
        activation='relu',
        kernel_regularizer=tf.keras.regularizers.l2(0.001) 
    ),
    # Zweiter Dropout-Layer
    tf.keras.layers.Dropout(0.4), 
    # Output-Layer: Anzahl der Neuronen entspricht der Anzahl der Klassen, Softmax für Wahrscheinlichkeiten
    tf.keras.layers.Dense(n_classes, activation='softmax')
])

# Optimierer und Kompilierung des Modells
optimizer = tf.keras.optimizers.Adam(learning_rate=0.005) # Adam-Optimierer mit spezifischer Lernrate
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy', # Verlustfunktion für Multi-Klassen-Klassifikation
              metrics=['accuracy']) # Metrik zur Überwachung des Trainings

model.summary() # Zeigt eine Zusammenfassung der Modellarchitektur

# --- Schritt 3.5: Definition von Callbacks und Klassengewichten ---
print("\n3.5 Definiere Callbacks und Klassengewichte...")
# ReduceLROnPlateau: Reduziert die Lernrate, wenn der Validierungsverlust stagniert
reduce_lr_callback = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6, verbose=1
)
# EarlyStopping: Stoppt das Training frühzeitig, wenn der Validierungsverlust nicht mehr sinkt
early_stopping_callback = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=12, restore_best_weights=True 
)

# Berechnung der Klassengewichte, um unausgewogene Klassenverteilungen zu berücksichtigen
class_weights_values = class_weight.compute_class_weight(
    'balanced', classes=np.unique(y_train), y=y_train
)
class_weights_dict = dict(enumerate(class_weights_values))
print(f"--> Verwende Klassengewichte: {class_weights_dict}")

# --- Schritt 3.6: Direktes Training des Modells ---
print("\n3.6 Starte direktes Modelltraining mit model.fit()...")
history = model.fit(
    X_train_scaled, # Trainingsdaten (Features)
    y_train, # Trainingslabels (Zielvariable)
    epochs=200, # Maximale Anzahl der Trainingsepochen
    batch_size=256, # Anzahl der Samples pro Gradienten-Update
    validation_data=(X_val_scaled, y_val), # Validierungsdaten zur Überwachung des Trainingsfortschritts
    callbacks=[early_stopping_callback, reduce_lr_callback], # Callbacks für das Training
    class_weight=class_weights_dict, # Gewichte für die Klassen
    verbose=1 # Zeigt den Fortschritt des Trainings an
)
print("--> Modelltraining abgeschlossen.")


# --- Schritt 3.7: Evaluation des trainierten Modells ---
def evaluate_and_plot_simple(X_data_scaled, y_data, dataset_name, model):
    """Eine Helfer-Funktion zur standardisierten Evaluation und Visualisierung für Neuronale Netze."""
    if X_data_scaled.size == 0 or y_data.empty:
        print(f"\n{dataset_name}-Datensatz ist leer. Evaluation wird übersprungen.")
        return None, None
    print("\n" + "-"*80); print(f" EVALUATION AUF {dataset_name.upper()}-DATEN"); print("-" * 80)
    
    # Mache Vorhersagen
    y_pred_proba = model.predict(X_data_scaled) # Vorhersage der Wahrscheinlichkeiten für jede Klasse
    y_pred = np.argmax(y_pred_proba, axis=1) # Wandelt Wahrscheinlichkeiten in Klassenvorhersagen um
    
    # Gebe den Klassifikations-Report aus
    class_names = list(class_mapping.keys())
    print("\nKlassifikations-Metriken:")
    print(classification_report(y_data, y_pred, labels=range(len(class_names)), target_names=class_names, zero_division=0))
    
    # Plotte die Konfusionsmatrix
    cm = confusion_matrix(y_data, y_pred, labels=range(len(class_names)))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    fig, ax = plt.subplots(figsize=(10, 8)); disp.plot(cmap=plt.cm.Blues, ax=ax, xticks_rotation='vertical'); ax.set_title(f'Klassifikationsmatrix für {dataset_name}'); plt.tight_layout(); plt.show()
    
    return y_pred, y_pred_proba

# Führe die Evaluation für Validierungs- und Test-Set durch
y_val_pred, y_val_pred_proba = evaluate_and_plot_simple(X_val_scaled, y_val, "Validation", model)
y_test_pred, y_test_pred_proba = evaluate_and_plot_simple(X_test_scaled, y_test, "Test", model)



# ... (Dein Code bis TEIL 3C bleibt unverändert) ...

print("\n" + "="*80); print(" TEIL 3C: MODELLINTERPRETATION MIT SHAP"); print("="*80)
try:
    # Wichtig: X_val_scaled ist bereits ein NumPy-Array hier.
    # Für SHAP ist es oft besser, wenn es ein Pandas DataFrame ist,
    # um Feature-Namen beizubehalten.
    # X_val_scaled_df wird weiter unten erstellt.

    if X_val_scaled.size > 0: # Prüft, ob Daten vorhanden sind

        # --- Vorbereitung der Daten für SHAP ---
        # Hintergrunddaten für den Explainer
        # `X_train_scaled` ist hier bereits ein NumPy-Array.
        # Konvertiere es zu DataFrame, um Feature-Namen für SHAP zu haben.
        # Wir können `features_in_model` aus Schritt 3.2 verwenden.
        X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=features_in_model)
        X_background = shap.sample(X_train_scaled_df, 100, random_state=42) # Feste random_state für Reproduzierbarkeit

        # Stichprobe der Validierungsdaten für die SHAP-Analyse
        X_val_scaled_df = pd.DataFrame(X_val_scaled, columns=features_in_model)
        X_val_shap_sample_df = shap.sample(X_val_scaled_df, 50, random_state=42) # Feste random_state

        print("\n--> Initialisiere SHAP-Explainer (DeepExplainer wird intern genutzt)...")
        # Für Keras/TensorFlow-Modelle ist shap.Explainer der empfohlene Weg.
        # Er leitet intern zu shap.DeepExplainer weiter, der für NN optimiert ist.
        explainer = shap.Explainer(model, X_background)

        print("\n--> Berechne SHAP-Werte für die Stichprobe der Validierungsdaten (kann einen Moment dauern)...")
        shap_values = explainer(X_val_shap_sample_df) # explainer(DataFrame) gibt ein shap.Explanation Objekt zurück

        # --- DEBUGGING-AUSGABEN HIER BEIBEHALTEN ---
        # shap_values ist hier ein shap.Explanation Objekt.
        # Die tatsächlichen SHAP-Werte sind in shap_values.values
        print(f"DEBUG: SHAP-Werte-Objekt-Typ: {type(shap_values)}")
        if hasattr(shap_values, 'values'):
            actual_shap_values = shap_values.values
            print(f"DEBUG: Typ der tatsächlichen SHAP-Werte: {type(actual_shap_values)}")
            if hasattr(actual_shap_values, 'shape'):
                print(f"DEBUG: Shape der tatsächlichen SHAP-Werte: {actual_shap_values.shape}")
                print(f"DEBUG: Dimensionen der tatsächlichen SHAP-Werte: {actual_shap_values.ndim}")
            # Auch hier kann es eine Liste von Arrays (selten für DeepExplainer) oder ein 3D Array sein
            if isinstance(actual_shap_values, list):
                print(f"DEBUG: tatsächliche SHAP-Werte sind Liste: True, Anzahl der Elemente: {len(actual_shap_values)}")
                if len(actual_shap_values) > 0:
                    if hasattr(actual_shap_values[0], 'shape'):
                        print(f"DEBUG: Shape des ersten Listenelements: {actual_shap_values[0].shape}")
            elif isinstance(actual_shap_values, np.ndarray):
                print(f"DEBUG: tatsächliche SHAP-Werte sind NumPy-Array.")
        # --- ENDE DEBUGGING-AUSGABEN ---

        MAX_SHAP_FEATURES = 40 # Angepasst auf 40, wie bei den Baummodellen

        # --- Globaler SHAP Feature Importance Plot (Bar-Diagramm) ---
        print("\n--> Erstelle SHAP Feature Importance Plot (global)...")
        
        # Die eigentlichen SHAP-Werte sind im .values-Attribut des Explanation-Objekts
        # Diese sind (Samples, Features, Classes) für Multiklassifikation
        if isinstance(shap_values.values, np.ndarray) and shap_values.values.ndim == 3:
            print("INFO: SHAP-Werte sind ein 3D NumPy-Array. Aggregiere für globalen Bar-Plot.")
            # Aggregiere über die Klassenachse (Achse 2)
            shap_values_for_global_plot = np.mean(np.abs(shap_values.values), axis=2) 
            
            # WICHTIG: shap.summary_plot erwartet hier das originale Explanation-Objekt
            # oder ein NumPy-Array, aber die Hintergrunddaten müssen passen.
            # Für den globalen Plot können wir das Explanation-Objekt verwenden,
            # aber nur mit dem Subset der Daten, für das die SHAP-Werte berechnet wurden.
            # Und wir müssen die .values entsprechend manipulieren, bevor wir sie übergeben.
            
            # Die einfachste Art ist, ein neues Explanation-Objekt für den Plot zu erstellen
            # mit den aggregierten Werten, wenn man den Bar-Plot über alle Klassen will.
            # Aber shap.summary_plot(ExplanationObject, plot_type="bar") kann das auch,
            # wenn das ExplanationObject multiklassifiziert ist.
            # Versuchen wir es mit dem rohen Array für den Bar Plot, das hat bei den Bäumen geklappt.
            shap.summary_plot(shap_values_for_global_plot, X_val_shap_sample_df, plot_type="bar", max_display=MAX_SHAP_FEATURES, show=False)
            plt.title("SHAP Feature Importance (Globaler Durchschnitt über alle Klassen) - Neuronales Netz")
            plt.tight_layout()
            plt.show()
            
            # --- SHAP Beeswarm Plots (Detailansicht) für jede Klasse ---
            print("\n--> Erstelle SHAP Beeswarm Plots (Detailansicht) für jede Klasse...")
            # Iterieren über die Klassen-Achse des 3D-Arrays: shap_values.values[:, :, i]
            for i, class_name in enumerate(list(class_mapping.keys())):
                print(f"    - Erstelle Beeswarm Plot für Klasse: '{class_name}' (Index {i}) - Neuronales Netz")
                plt.figure(figsize=(12, 8))
                # shap.summary_plot erwartet hier die SHAP-Werte für die spezifische Klasse (2D Array)
                # und die entsprechenden Hintergrunddaten (X_val_shap_sample_df)
                shap.summary_plot(shap_values.values[:, :, i], X_val_shap_sample_df, max_display=MAX_SHAP_FEATURES, feature_names=features_in_model, show=False)
                plt.title(f"SHAP-Werte für Klasse: '{class_name}' - Neuronales Netz")
                plt.tight_layout()
                plt.show()
                
        elif isinstance(shap_values.values, list) and len(shap_values.values) > 1: # Falls es doch eine Liste von Arrays ist (unwahrscheinlich für DeepExplainer)
            print("INFO: SHAP-Werte sind Liste für Multiklasse. Aggregiere für globalen Bar-Plot.")
            shap_values_for_global_plot = np.mean(np.abs(np.array(shap_values.values)), axis=0) # Hier ist axis=0 korrekt
            
            shap.summary_plot(shap_values_for_global_plot, X_val_shap_sample_df, plot_type="bar", max_display=MAX_SHAP_FEATURES, show=False)
            plt.title("SHAP Feature Importance (Globaler Durchschnitt über alle Klassen) - Neuronales Netz")
            plt.tight_layout()
            plt.show()
            
            # Beeswarm Plots pro Klasse aus der Liste
            print("\n--> Erstelle SHAP Beeswarm Plots (Detailansicht) für jede Klasse...")
            if len(shap_values.values) == len(class_mapping):
                for i, class_name in enumerate(list(class_mapping.keys())):
                    print(f"    - Erstelle Beeswarm Plot für Klasse: '{class_name}' (Index {i}) - Neuronales Netz")
                    plt.figure(figsize=(12, 8))
                    shap.summary_plot(shap_values.values[i], X_val_shap_sample_df, max_display=MAX_SHAP_FEATURES, feature_names=features_in_model, show=False)
                    plt.title(f"SHAP-Werte für Klasse: '{class_name}' - Neuronales Netz")
                    plt.tight_layout()
                    plt.show()
            else:
                print(f"WARNUNG: Anzahl der SHAP-Werte-Listen ({len(shap_values.values)}) stimmt nicht mit der Anzahl der Klassen ({len(class_mapping)}) überein. Überspringe Beeswarm Plots pro Klasse.")
                
        else: # Dies wird ausgeführt, wenn shap_values.values weder 3D Array noch Liste > 1 Element ist
            print("INFO: SHAP-Werte sind kein Multiklassen-Format (oder nur eine Klasse). Erstelle einen einzelnen globalen Beeswarm/Bar-Plot.")
            # Für den globalen Plot bei nur einer Klasse, Beeswarm ist oft aussagekräftiger
            # Hier müssen wir shap_values.values übergeben (das sollte ein 2D-Array sein)
            shap.summary_plot(shap_values.values, X_val_shap_sample_df, max_display=MAX_SHAP_FEATURES, feature_names=features_in_model, show=False)
            plt.title("SHAP-Werte (Gesamt / Binäre Klassifikation) - Neuronales Netz")
            plt.tight_layout()
            plt.show()
            print("WARNUNG: Keine Beeswarm Plots pro Klasse möglich, da SHAP-Werte nicht im Multiklassen-Listenformat vorliegen.")
            
    else:
        print("Validierungsdaten sind leer, SHAP-Analyse wird übersprungen.")
except Exception as e:
    print(f"Fehler bei der Modell-Interpretation: {e}")
    import traceback
    traceback.print_exc()

print("\n[SUCCESS] Teil 3 abgeschlossen.")

# --- TEIL 4: SPEICHERN DER ERGEBNISSE ---
print("\n" + "="*80); print(" TEIL 4: SPEICHERN DER ERGEBNISSE"); print("=" * 80)
if y_test_pred is not None:
    test_results_df = X_test_raw.copy()
    test_results_df[label_column_ml] = y_test
    test_results_df['prediction'] = y_test_pred
    proba_df = pd.DataFrame(y_test_pred_proba, columns=[f'prediction_proba_{cls}' for cls in class_mapping.keys()], index=test_results_df.index)
    test_results_df = pd.concat([test_results_df, proba_df], axis=1)
    
    test_results_df.to_csv("test_data_final_classification_NEURAL_NET.csv", index=True)
    print(f"--> Testdaten mit Vorhersagen und Wahrscheinlichkeiten gespeichert: test_data_final_classification_NEURAL_NET.csv")
else:
    print("--> Testdaten sind leer, keine Speicherung.")

print("\nSkript vollständig abgeschlossen.")
```

```python
# --- TEIL 0: Bibliotheken importieren und Setup ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pastas as ps
import pyarrow.dataset as ds
import tensorflow as tf # Spezifisch für Neuronale Netze (Keras/TensorFlow)
from sklearn.utils import class_weight # Für die Handhabung ungleicher Klassenverteilungen
import shap # Für die Modell-Interpretation (SHapley Additive exPlanations)
from sklearn.preprocessing import StandardScaler # Zur Skalierung numerischer Features
from sklearn.impute import SimpleImputer # Zum Füllen von fehlenden Werten
import gc # Für die manuelle Speicherbereinigung
from statsmodels.tsa.seasonal import STL # Für die saisonale Zerlegung von Zeitreihen
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay # Zur Evaluation des Modells
from tqdm import tqdm # Für schöne Fortschrittsbalken bei langen Schleifen
from contextlib import redirect_stdout # Um störende Ausgaben von Bibliotheken zu unterdrücken
import io # Für die Umleitung von Ausgaben

# NEU: Import für KerasClassifier
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV # GridSearchCV oder RandomizedSearchCV

# --- Konfiguration ---
# Loglevel von 'pastas' reduzieren, um die Konsole sauber zu halten
ps.set_log_level("ERROR")
print("Starte Skript: Klassifikations-Pipeline mit NEURONALEM NETZ")

# Setzen von Random Seeds für Reproduzierbarkeit
np.random.seed(42)
tf.random.set_seed(42)


# --- Definition der Features (UNVERÄNDERT) ---
# Hier werden die Spaltennamen für verschiedene Feature-Typen zentral definiert.
STATIC_CATEGORICAL_FEATURES = ["main_landuse"]
STATIC_NUMERICAL_FEATURES = ["ground_elevation_m_asl", "topographic_slope_degrees"]
DYNAMIC_NUMERICAL_FEATURES_RAW = [
    "groundwater_level_m_asl", "precipitation_gpcc_mm_year-1",
    "potential_evapotranspiration_era5_mm_year-1", "air_temperature_°C", "snow_depth_m",
    "ndvi_ratio", "lai_low_vegetation_ratio", "lai_high_vegetation_ratio",
    "withdrawal_industrial_m3_year-1", "withdrawal_domestic_m3_year-1"
]

# --- TEIL 1 & 2: DATEN LADEN UND FEATURE ENGINEERING ---
print("\n" + "="*80); print(" TEIL 1 & 2: DATEN LADEN UND FEATURE ENGINEERING"); print("="*80)

# --- Schritt 1.1: Statische Attribute laden und filtern ---
print("1.1 Lade statische Attribute und filtere relevante Stationen...")
path_attributes = "C:/MLmitPython/ML-Python/Daten/grow_attributes.parquet"
path_timeseries = "C:/MLmitPython/ML-Python/Daten/grow_timeseries.parquet"
try:
    attr_df_full = pd.read_parquet(path_attributes)
except FileNotFoundError: raise

# Filtere die Attribute nach den gewünschten Kriterien (Tschechien, lange Zeitreihen, keine Lücken etc.)
attr_filtered_df = attr_df_full[(attr_df_full["country"] == "CZE") & (attr_df_full["length_years"] >= 30) & (attr_df_full["gap_fraction"] == 0.0) & (pd.to_datetime(attr_df_full["ending_date"]) >= pd.Timestamp("2023-12-01"))]
if attr_filtered_df.empty: raise ValueError("Keine passende Zeitreihe für CZ für ML gefunden.")
cz_ids_for_ml = attr_filtered_df["GROW_ID"].dropna().unique().tolist()
print(f"--> {len(cz_ids_for_ml)} relevante Stationen für die Analyse gefunden.")

# --- Schritt 1.2: Dynamische Zeitreihen laden ---
print("\n1.2 Lade dynamische Zeitreihen für die ausgewählten Stationen...")
dynamic_feature_columns_to_load = ["GROW_ID", "date"] + DYNAMIC_NUMERICAL_FEATURES_RAW
try:
    # PyArrow wird für schnelles, speichereffizientes Laden der großen Parquet-Datei verwendet
    timeseries_dataset_raw = ds.dataset(path_timeseries, format="parquet")
    available_columns_in_parquet = timeseries_dataset_raw.schema.names
    columns_to_actually_load = [col for col in dynamic_feature_columns_to_load if col in available_columns_in_parquet]
    timeseries_table_raw = timeseries_dataset_raw.to_table(filter=ds.field("GROW_ID").isin(cz_ids_for_ml), columns=columns_to_actually_load)
    raw_ts_df_full = timeseries_table_raw.to_pandas(self_destruct=True, split_blocks=True, zero_copy_only=False)
except FileNotFoundError: raise
raw_ts_df_full["date"] = pd.to_datetime(raw_ts_df_full["date"])
print("--> Zeitreihen erfolgreich geladen.")

# --- Schritt 1.3: Erste Feature-Anpassungen ---
print("\n1.3 Berechne monatliche Features aus jährlichen Werten...")
if "precipitation_gpcc_mm_year-1" in raw_ts_df_full.columns: raw_ts_df_full["precipitation_gpcc_mm_month-1"] = (raw_ts_df_full["precipitation_gpcc_mm_year-1"] / 12)
if "potential_evapotranspiration_era5_mm_year-1" in raw_ts_df_full.columns: raw_ts_df_full["potential_evapotranspiration_era5_mm_month-1"] = (raw_ts_df_full["potential_evapotranspiration_era5_mm_year-1"] / 12)

# Gruppiere die Zeitreihen nach ihrer ID für die Einzelverarbeitung
grouped_ts_data = raw_ts_df_full.set_index('date').groupby('GROW_ID'); del raw_ts_df_full; gc.collect()

# --- Schritt 1.4: Haupt-Feature-Engineering pro Station ---
print("\n1.4 Führe SGI-Berechnung und Feature-Engineering pro Station durch...")
# Definiere eine Funktion, die alle Schritte für eine einzelne Station durchführt
def process_grow_id_data(grow_id_data):
    # Resample Grundwasserstand auf monatliche Werte
    gwl_series_raw = grow_id_data["groundwater_level_m_asl"].dropna()
    if gwl_series_raw.empty or len(gwl_series_raw.resample('MS').mean().dropna()) < 120: return None # Mindestens 10 Jahre Daten benötigt
    gwl_series_monthly = gwl_series_raw.resample('MS').mean()
    
    # Berechne den Standardized Groundwater Index (SGI) mit der 'pastas'-Bibliothek
    try:
        with redirect_stdout(io.StringIO()): # Unterdrückt Konsolenausgaben von pastas
            sgi_observed_monthly = ps.stats.sgi(gwl_series_monthly)
        if sgi_observed_monthly.empty or sgi_observed_monthly.isnull().all(): return None
    except Exception: return None
    
    # Zerlege die SGI-Zeitreihe in Trend, Saison und Residuum (Rest)
    sgi_trend, sgi_seasonal, sgi_resid = (pd.Series(dtype=np.float32), pd.Series(dtype=np.float32), pd.Series(dtype=np.float32))
    sgi_for_stl = sgi_observed_monthly.dropna()
    if len(sgi_for_stl) >= 25: # STL benötigt eine minimale Anzahl an Datenpunkten
        try:
            stl = STL(sgi_for_stl, period=12, robust=True); res_stl = stl.fit()
            sgi_trend, sgi_seasonal, sgi_resid = (res_stl.trend.reindex(sgi_observed_monthly.index), res_stl.seasonal.reindex(sgi_observed_monthly.index), res_stl.resid.reindex(sgi_observed_monthly.index))
        except Exception: pass
        
    # Sammle alle berechneten und originalen Features in einem Dictionary
    monthly_features_dict = {'sgi_label': sgi_observed_monthly, 'sgi_trend': sgi_trend, 'sgi_seasonal': sgi_seasonal, 'sgi_resid': sgi_resid}
    dynamic_map = {"precipitation_gpcc_mm_month-1": "precipitation_monthly", "potential_evapotranspiration_era5_mm_year-1": "evapotranspiration_monthly", "air_temperature_°C": "air_temperature_monthly", "snow_depth_m": "snow_depth_monthly", "ndvi_ratio": "ndvi_monthly", "lai_low_vegetation_ratio": "lai_low_veg_monthly", "lai_high_vegetation_ratio": "lai_high_veg_monthly"}
    for orig, target in dynamic_map.items(): monthly_features_dict[target] = grow_id_data[orig].dropna().resample('MS').mean().reindex(sgi_observed_monthly.index) if orig in grow_id_data.columns else pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    yearly_map = {"withdrawal_industrial_m3_year-1": "withdrawal_industrial_monthly", "withdrawal_domestic_m3_year-1": "withdrawal_domestic_monthly"}
    for orig, target in yearly_map.items():
        if orig in grow_id_data.columns:
            series_yearly = grow_id_data[orig].dropna(); series_yearly = series_yearly[~series_yearly.index.duplicated(keep='first')] if not series_yearly.index.is_unique else series_yearly; monthly_features_dict[target] = series_yearly.resample('MS').ffill().reindex(sgi_observed_monthly.index)
        else: monthly_features_dict[target] = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
        
    return pd.DataFrame(monthly_features_dict).dropna(subset=['sgi_label'])

# Wende die Funktion auf alle Stations-Gruppen an und zeige einen Fortschrittsbalken
processed_groups = []
for grow_id, group_data in tqdm(grouped_ts_data, desc="SGI-Berechnung & Dekomposition"):
    processed_df = process_grow_id_data(group_data)
    if processed_df is not None and not processed_df.empty: processed_df['GROW_ID'] = grow_id; processed_groups.append(processed_df)
if not processed_groups: raise ValueError("Keine Daten für ML-Modell nach Verarbeitung.")

# --- Schritt 1.5: Finale Datenzusammenführung und Feature-Erstellung ---
print("\n1.5 Führe dynamische und statische Features zusammen...")
df_ml_input_dynamic = pd.concat(processed_groups); del processed_groups; gc.collect()
static_features_to_select = ["GROW_ID"] + STATIC_NUMERICAL_FEATURES + STATIC_CATEGORICAL_FEATURES
static_features_to_select = [col for col in static_features_to_select if col in attr_df_full.columns]
attr_df_ml_static = attr_df_full[attr_df_full['GROW_ID'].isin(cz_ids_for_ml)][static_features_to_select].copy().drop_duplicates(subset=['GROW_ID'], keep='first')
df_ml_input_dynamic.reset_index(inplace=True)
df_ml_input = pd.merge(df_ml_input_dynamic, attr_df_ml_static, on="GROW_ID", how="left").set_index('date')
del df_ml_input_dynamic, attr_df_ml_static, attr_df_full, attr_filtered_df; gc.collect()
df_ml_processed = df_ml_input.copy(); del df_ml_input; gc.collect()

print("\n1.6 Erstelle zeitversetzte (Lag) Features und Zielvariable...")
# Erstelle Lag-Features für die SGI-Komponenten (Werte von vor 1, 3, 6, 12 Monaten)
for lag_col_base in ['sgi_label', 'sgi_trend', 'sgi_seasonal', 'sgi_resid']:
    if lag_col_base in df_ml_processed.columns:
        for lag_val in [1, 3, 6, 12]: df_ml_processed[f'{lag_col_base}_lag_{lag_val}M'] = df_ml_processed.groupby('GROW_ID')[lag_col_base].shift(lag_val)

df_ml_processed['month_num'] = df_ml_processed.index.month.astype(np.int8)
df_ml_processed['year'] = df_ml_processed.index.year.astype(np.int32)

# Erstelle die Zielvariable: der SGI-Wert 6 Monate in der Zukunft
VORHERSAGE_HORIZONT_MONATE = 6
sgi_target_col_continuous = f'sgi_target_{VORHERSAGE_HORIZONT_MONATE}M_future'
df_ml_processed[sgi_target_col_continuous] = df_ml_processed.groupby('GROW_ID')['sgi_label'].shift(-VORHERSAGE_HORIZONT_MONATE)

# --- Schritt 1.7: Zielvariable in Klassen umwandeln und Features vorbereiten ---
print("\n1.7 Wandle Zielvariable in Klassen um und kodiere kategorische Features...")
def sgi_to_drought_class(z_score):
    if pd.isna(z_score): return None
    if z_score > -0.84: return 'Normal'
    if -1.50 <= z_score <= -0.84: return 'Mild_Moderate'
    if -2.00 <= z_score <= -1.51: return 'Severe'
    return 'Extreme'
df_ml_processed['drought_class_str'] = df_ml_processed[sgi_target_col_continuous].apply(sgi_to_drought_class)
class_mapping = {'Normal': 0, 'Mild_Moderate': 1, 'Severe': 2, 'Extreme': 3}
df_ml_processed['drought_class_target'] = df_ml_processed['drought_class_str'].map(class_mapping).astype('Int64')
label_column_ml = 'drought_class_target'

# Definiere die zu kodierenden Spalten dynamisch, basierend auf dem, was wirklich im DataFrame ist
print("\n1.7.1 Kodierung der kategorischen Features...")
categorical_cols_to_encode = [col for col in STATIC_CATEGORICAL_FEATURES if col in df_ml_processed.columns] + ['month_num']
print(f"--> Folgende kategorische Spalten werden kodiert: {categorical_cols_to_encode}")

# Die Schleife bleibt gleich, verarbeitet aber nur existierende Spalten
dummy_encoded_cols_list = []
for col_to_encode in categorical_cols_to_encode:
    print(f"      - Kodiere '{col_to_encode}'...")
    dummies = pd.get_dummies(df_ml_processed[col_to_encode], prefix=col_to_encode, dummy_na=False, dtype=np.int8)
    df_ml_processed = pd.concat([df_ml_processed, dummies], axis=1); df_ml_processed.drop(columns=[col_to_encode], inplace=True)
    dummy_encoded_cols_list.extend(dummies.columns.tolist())

# --- Schritt 1.8: Finale Datenbereinigung ---
print("\n1.8 Bereinige finale Datensätze...")
# Fülle verbleibende Lücken in numerischen Spalten mit dem Median
numerical_static_cols_for_imputation = [col for col in STATIC_NUMERICAL_FEATURES if col in df_ml_processed.columns]
if numerical_static_cols_for_imputation:
    df_ml_processed[numerical_static_cols_for_imputation] = SimpleImputer(strategy='median').fit_transform(df_ml_processed[numerical_static_cols_for_imputation])

# Definiere die finale Liste aller Features für das Modell
base_features_ml = ["precipitation_monthly", "evapotranspiration_monthly", "air_temperature_monthly", "snow_depth_monthly", "ndvi_monthly", "lai_low_veg_monthly", "lai_high_veg_monthly", "withdrawal_industrial_monthly", "withdrawal_domestic_monthly"]
lag_features_ml = [col for col in df_ml_processed.columns if '_lag_' in col]
current_sgi_features = ['sgi_label', 'sgi_trend', 'sgi_seasonal', 'sgi_resid']
features_columns_ml = list(set([col for col in base_features_ml if col in df_ml_processed.columns] + lag_features_ml + current_sgi_features + numerical_static_cols_for_imputation + dummy_encoded_cols_list))
# Entferne Zeilen, in denen wichtige Features oder das Label fehlen
columns_to_check_for_dropna_smart = [label_column_ml, 'sgi_label_lag_1M', 'precipitation_monthly']
df_ml_processed.dropna(subset=[col for col in columns_to_check_for_dropna_smart if col in df_ml_processed.columns], inplace=True)
df_ml_processed[label_column_ml] = df_ml_processed[label_column_ml].astype(int)

# Finale Aufteilung in X (Features) und y (Zielvariable)
X_all_features_raw = df_ml_processed[features_columns_ml].copy()
y_all_labels = df_ml_processed[label_column_ml].copy()
X_meta_for_split = df_ml_processed[['GROW_ID', 'year']].copy()
del df_ml_processed; gc.collect()
print("\n[SUCCESS] Teil 1 & 2 abgeschlossen.")


# --- TEIL 3: MODELLTRAINING UND EVALUATION ---
print("\n" + "="*80); print(" TEIL 3: MODELLTRAINING & EVALUATION (NEURONALES NETZ)"); print("="*80)

# --- Schritt 3.1: Zeitbasierter Train-Validation-Test Split ---
temp_year_series = X_meta_for_split['year']
unique_years_ml = np.sort(temp_year_series.unique())
split_year_1 = int(np.percentile(unique_years_ml, 80))
split_year_2 = int(np.percentile(unique_years_ml, 90))
print(f"--> Erstelle zeitbasierten 80/10/10 Split:")
print(f"    - Trainingsdaten: Jahre < {split_year_1} (80%)")
print(f"    - Validierungsdaten: Jahre >= {split_year_1} und < {split_year_2} (10%)")
print(f"    - Testdaten: Jahre >= {split_year_2} (10%)")
train_mask = temp_year_series < split_year_1
val_mask = (temp_year_series >= split_year_1) & (temp_year_series < split_year_2)
test_mask = temp_year_series >= split_year_2
X_train_raw, X_val_raw, X_test_raw = X_all_features_raw[train_mask], X_all_features_raw[val_mask], X_all_features_raw[test_mask]
y_train, y_val, y_test = y_all_labels[train_mask], y_all_labels[val_mask], y_all_labels[test_mask]
print(f"\nGröße der Datensätze:")
print(f"    - Training:   {X_train_raw.shape[0]} Zeilen")
print(f"    - Validation: {X_val_raw.shape[0]} Zeilen")
print(f"    - Test:       {X_test_raw.shape[0]} Zeilen")
del X_all_features_raw, y_all_labels, X_meta_for_split; gc.collect()


# --- Schritt 3.2: Finale Feature-Übersicht ---
print("\n" + "-"*80)
features_in_model = sorted(X_train_raw.columns.tolist())
print(f"--> Folgende {len(features_in_model)} Features werden für das Modell verwendet:")
print("-" * 80)
max_len = max(len(f) for f in features_in_model) + 2
num_columns = 4
for i in range(0, len(features_in_model), num_columns):
    row = [f"{feat:<{max_len}}" for feat in features_in_model[i:i+num_columns]]
    print("".join(row))
print("-" * 80)


# --- Schritt 3.3: Manuelles Preprocessing der Daten (vor dem NN-Training) ---
print("\n3.3 Führe manuelles Preprocessing durch (Imputer & Scaler)...")
# Imputer zum Füllen fehlender Werte (z.B. mit dem Mittelwert)
imputer = SimpleImputer(strategy='mean')
# Scaler zum Skalieren der Daten (z.B. auf Mittelwert 0 und Standardabweichung 1)
scaler = StandardScaler()

# Daten transformieren: zuerst Imputation, dann Skalierung
X_train_scaled = scaler.fit_transform(imputer.fit_transform(X_train_raw))
X_val_scaled = scaler.transform(imputer.transform(X_val_raw))
X_test_scaled = scaler.transform(imputer.transform(X_test_raw))
print("--> Daten erfolgreich imputiert und skaliert.")


# --- Schritt 3.4: Definition des Keras-Modells (Neuronales Netz) als Funktion für GridSearchCV ---
print("\n3.4 Definiere die Architektur des Neuronalen Netzes als Funktion für Hyperparameter-Tuning...")

def create_nn_model(n_features, n_classes, num_hidden_layers=2, neurons_per_layer=(80, 40), dropout_rate=0.4, l2_reg=0.001, learning_rate=0.005):
    """
    Erstellt ein Keras Sequential Modell mit konfigurierbaren Hyperparametern.
    """
    if isinstance(neurons_per_layer, int): # Falls nur eine Zahl für Neuronen übergeben wird
        neurons_per_layer = (neurons_per_layer,) * num_hidden_layers
    elif len(neurons_per_layer) != num_hidden_layers:
        raise ValueError("Anzahl der Neuronen in 'neurons_per_layer' muss der 'num_hidden_layers' entsprechen.")

    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(n_features,)),
    ])
    
    for i in range(num_hidden_layers):
        model.add(tf.keras.layers.Dense(
            neurons_per_layer[i],
            activation='relu',
            kernel_regularizer=tf.keras.regularizers.l2(l2_reg)
        ))
        model.add(tf.keras.layers.Dropout(dropout_rate))
    
    model.add(tf.keras.layers.Dense(n_classes, activation='softmax'))

    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy']) # accuracy ist für Tuning in KerasClassifier ok, da f1_weighted danach kommt
    return model

# Anzahl der Features und Klassen für das Modell
n_features = X_train_scaled.shape[1]
n_classes = len(class_mapping)

# Berechne Klassengewichte für das Training
class_weights_values = class_weight.compute_class_weight(
    'balanced', classes=np.unique(y_train), y=y_train
)
class_weights_dict = dict(enumerate(class_weights_values))
print(f"--> Verwende Klassengewichte: {class_weights_dict}")

# ... (Ihr Code bis zum param_grid_nn_reduced)

print("\n3.5 Starte Hyperparameter-Tuning für das Neuronale Netz mit GridSearchCV...")

nn_classifier = KerasClassifier(
    model=create_nn_model,
    n_features=n_features,
    n_classes=n_classes,
    epochs=100,
    batch_size=128,
    verbose=0,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6, verbose=0),
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True, verbose=0)
    ]
)

# Definiere separate Parameter-Grids für jede Anzahl von Hidden Layers
# Grid 1: Eine Hidden Layer
param_grid_nn_1_layer = {
    'model__num_hidden_layers': [1],
    'model__neurons_per_layer': [(64,), (128,)], # Hier passen die Tupel zur Anzahl der Schichten
    'model__dropout_rate': [0.3],
    'model__l2_reg': [0.001],
    'model__learning_rate': [0.001, 0.005],
    'batch_size': [128, 256]
}
# Fits: 1 * 2 * 1 * 1 * 2 * 2 = 8 Kombinationen * 5 CV = 40 Fits

# Grid 2: Zwei Hidden Layers
param_grid_nn_2_layers = {
    'model__num_hidden_layers': [2],
    'model__neurons_per_layer': [(64, 32), (128, 64)], # Hier passen die Tupel zur Anzahl der Schichten
    'model__dropout_rate': [0.3],
    'model__l2_reg': [0.001],
    'model__learning_rate': [0.001, 0.005],
    'batch_size': [128, 256]
}
# Fits: 1 * 2 * 1 * 1 * 2 * 2 = 8 Kombinationen * 5 CV = 40 Fits

# Kombiniere die Grids (GridSearchCV kann eine Liste von Parameter-Grids verarbeiten)
param_grid_nn_combined = [
    param_grid_nn_1_layer,
    param_grid_nn_2_layers
]

# Da wir spezifische Kombinationen definieren, ist GridSearchCV hier gut.
# Aber RandomizedSearchCV ist immer noch eine gute Option für größere Suchräume
# und kann auch eine Liste von Grids verarbeiten.
use_randomized_search = True # Oder False für GridSearchCV mit combined grid

if use_randomized_search:
    grid_search_nn = RandomizedSearchCV(
        estimator=nn_classifier,
        param_distributions=param_grid_nn_combined, # Jetzt Liste von Grids
        n_iter=20, # Teste z.B. 20 zufällige Kombinationen aus dem Gesamt-Raum
        cv=5,
        scoring='f1_weighted',
        random_state=42,
        verbose=2,
        n_jobs=-1,
        error_score='raise'
    )
    print(f"\nStarte Randomisierte Suche für Neuronales Netz (testet bis zu {grid_search_nn.n_iter} Kombinationen)...")
else:
    grid_search_nn = GridSearchCV(
        estimator=nn_classifier,
        param_grid=param_grid_nn_combined, # Jetzt Liste von Grids
        cv=5,
        scoring='f1_weighted',
        verbose=2,
        n_jobs=-1,
        error_score='raise'
    )
    print(f"\nStarte Gitter-Suche für Neuronales Netz (testet {len(list(param_grid_nn_1_layer.values())[0])*len(list(param_grid_nn_1_layer.values())[1])*len(list(param_grid_nn_1_layer.values())[2])*len(list(param_grid_nn_1_layer.values())[3])*len(list(param_grid_nn_1_layer.values())[4])*len(list(param_grid_nn_1_layer.values())[5]) + len(list(param_grid_nn_2_layers.values())[0])*len(list(param_grid_nn_2_layers.values())[1])*len(list(param_grid_nn_2_layers.values())[2])*len(list(param_grid_nn_2_layers.values())[3])*len(list(param_grid_nn_2_layers.values())[4])*len(list(param_grid_nn_2_layers.values())[5])} Kombinationen)...")


grid_search_nn.fit(X_train_scaled, y_train, class_weight=class_weights_dict) # class_weight hier übergeben

best_nn_model = grid_search_nn.best_estimator_
print(f"\nNeuronales Netz Beste Parameter: {grid_search_nn.best_params_}")
print(f"Bester gewichteter F1-Score (CV): {grid_search_nn.best_score_:.4f}")

# --- Schritt 3.6: Evaluation des trainierten Modells (Bestes Modell aus Grid Search) ---
def evaluate_and_plot_simple(X_data_scaled, y_data, dataset_name, model):
    """Eine Helfer-Funktion zur standardisierten Evaluation und Visualisierung für Neuronale Netze."""
    if X_data_scaled.size == 0 or y_data.empty:
        print(f"\n{dataset_name}-Datensatz ist leer. Evaluation wird übersprungen.")
        return None, None
    print("\n" + "-"*80); print(f" EVALUATION AUF {dataset_name.upper()}-DATEN"); print("-" * 80)
    
    # Mache Vorhersagen
    # predict_proba ist in KerasClassifier automatisch verfügbar
    y_pred_proba = model.predict_proba(X_data_scaled) 
    y_pred = np.argmax(y_pred_proba, axis=1)
    
    # Gebe den Klassifikations-Report aus
    class_names = list(class_mapping.keys())
    print("\nKlassifikations-Metriken:")
    print(classification_report(y_data, y_pred, labels=range(len(class_names)), target_names=class_names, zero_division=0))
    
    # Plotte die Konfusionsmatrix
    cm = confusion_matrix(y_data, y_pred, labels=range(len(class_names)))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    fig, ax = plt.subplots(figsize=(10, 8)); disp.plot(cmap=plt.cm.Blues, ax=ax, xticks_rotation='vertical'); ax.set_title(f'Klassifikationsmatrix für {dataset_name}'); plt.tight_layout(); plt.show()
    
    return y_pred, y_pred_proba

# Führe die Evaluation für Validierungs- und Test-Set durch mit dem besten Modell
y_val_pred, y_val_pred_proba = evaluate_and_plot_simple(X_val_scaled, y_val, "Validation", best_nn_model)
y_test_pred, y_test_pred_proba = evaluate_and_plot_simple(X_test_scaled, y_test, "Test", best_nn_model)


# --- TEIL 3C: MODELLINTERPRETATION MIT SHAP ---
print("\n" + "="*80); print(" TEIL 3C: MODELLINTERPRETATION MIT SHAP"); print("="*80)
try:
    # Wichtig: X_val_scaled ist bereits ein NumPy-Array hier.
    # Für SHAP ist es oft besser, wenn es ein Pandas DataFrame ist,
    # um Feature-Namen beizubehalten.
    # X_val_scaled_df wird weiter unten erstellt.

    if X_val_scaled.size > 0: # Prüft, ob Daten vorhanden sind

        # --- Vorbereitung der Daten für SHAP ---
        # Hintergrunddaten für den Explainer
        # `X_train_scaled` ist hier bereits ein NumPy-Array.
        # Konvertiere es zu DataFrame, um Feature-Namen für SHAP zu haben.
        # Wir können `features_in_model` aus Schritt 3.2 verwenden.
        X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=features_in_model)
        X_background = shap.sample(X_train_scaled_df, 100, random_state=42) # Feste random_state für Reproduzierbarkeit

        # Stichprobe der Validierungsdaten für die SHAP-Analyse
        X_val_scaled_df = pd.DataFrame(X_val_scaled, columns=features_in_model)
        X_val_shap_sample_df = shap.sample(X_val_scaled_df, 50, random_state=42) # Feste random_state

        print("\n--> Initialisiere SHAP-Explainer (DeepExplainer wird intern genutzt)...")
        # Für Keras/TensorFlow-Modelle ist shap.Explainer der empfohlene Weg.
        # Er leitet intern zu shap.DeepExplainer weiter, der für NN optimiert ist.
        # Da best_nn_model jetzt ein KerasClassifier ist, müssen wir sein internes Keras-Modell extrahieren
        explainer = shap.Explainer(best_nn_model.model_, X_background) # <- HIER DIE ANPASSUNG!

        print("\n--> Berechne SHAP-Werte für die Stichprobe der Validierungsdaten (kann einen Moment dauern)...")
        shap_values = explainer(X_val_shap_sample_df) # explainer(DataFrame) gibt ein shap.Explanation Objekt zurück

        # --- DEBUGGING-AUSGABEN HIER BEIBEHALTEN ---
        # shap_values ist hier ein shap.Explanation Objekt.
        # Die tatsächlichen SHAP-Werte sind in shap_values.values
        print(f"DEBUG: SHAP-Werte-Objekt-Typ: {type(shap_values)}")
        if hasattr(shap_values, 'values'):
            actual_shap_values = shap_values.values
            print(f"DEBUG: Typ der tatsächlichen SHAP-Werte: {type(actual_shap_values)}")
            if hasattr(actual_shap_values, 'shape'):
                print(f"DEBUG: Shape der tatsächlichen SHAP-Werte: {actual_shap_values.shape}")
                print(f"DEBUG: Dimensionen der tatsächlichen SHAP-Werte: {actual_shap_values.ndim}")
            # Auch hier kann es eine Liste von Arrays (selten für DeepExplainer) oder ein 3D Array sein
            if isinstance(actual_shap_values, list):
                print(f"DEBUG: tatsächliche SHAP-Werte sind Liste: True, Anzahl der Elemente: {len(actual_shap_values)}")
                if len(actual_shap_values) > 0:
                    if hasattr(actual_shap_values[0], 'shape'):
                        print(f"DEBUG: Shape des ersten Listenelements: {actual_shap_values[0].shape}")
            elif isinstance(actual_shap_values, np.ndarray):
                print(f"DEBUG: tatsächliche SHAP-Werte sind NumPy-Array.")
        # --- ENDE DEBUGGING-AUSGABEN ---

        MAX_SHAP_FEATURES = 40 # Angepasst auf 40, wie bei den Baummodellen

        # --- Globaler SHAP Feature Importance Plot (Bar-Diagramm) ---
        print("\n--> Erstelle SHAP Feature Importance Plot (global)...")
        
        # Die eigentlichen SHAP-Werte sind im .values-Attribut des Explanation-Objekts
        # Diese sind (Samples, Features, Classes) für Multiklassifikation
        if isinstance(shap_values.values, np.ndarray) and shap_values.values.ndim == 3:
            print("INFO: SHAP-Werte sind ein 3D NumPy-Array. Aggregiere für globalen Bar-Plot.")
            # Aggregiere über die Klassenachse (Achse 2)
            shap_values_for_global_plot = np.mean(np.abs(shap_values.values), axis=2) 
            
            # WICHTIG: shap.summary_plot erwartet hier das originale Explanation-Objekt
            # oder ein NumPy-Array, aber die Hintergrunddaten müssen passen.
            # Für den globalen Plot können wir das Explanation-Objekt verwenden,
            # aber nur mit dem Subset der Daten, für das die SHAP-Werte berechnet wurden.
            # Und wir müssen die .values entsprechend manipulieren, bevor wir sie übergeben.
            
            # Die einfachste Art ist, ein neues Explanation-Objekt für den Plot zu erstellen
            # mit den aggregierten Werten, wenn man den Bar-Plot über alle Klassen will.
            # Aber shap.summary_plot(ExplanationObject, plot_type="bar") kann das auch,
            # wenn das ExplanationObject multiklassifiziert ist.
            # Versuchen wir es mit dem rohen Array für den Bar Plot, das hat bei den Bäumen geklappt.
            shap.summary_plot(shap_values_for_global_plot, X_val_shap_sample_df, plot_type="bar", max_display=MAX_SHAP_FEATURES, show=False)
            plt.title("SHAP Feature Importance (Globaler Durchschnitt über alle Klassen) - Neuronales Netz")
            plt.tight_layout()
            plt.show()
            
            # --- SHAP Beeswarm Plots (Detailansicht) für jede Klasse ---
            print("\n--> Erstelle SHAP Beeswarm Plots (Detailansicht) für jede Klasse...")
            # Iterieren über die Klassen-Achse des 3D-Arrays: shap_values.values[:, :, i]
            for i, class_name in enumerate(list(class_mapping.keys())):
                print(f"    - Erstelle Beeswarm Plot für Klasse: '{class_name}' (Index {i}) - Neuronales Netz")
                plt.figure(figsize=(12, 8))
                # shap.summary_plot erwartet hier die SHAP-Werte für die spezifische Klasse (2D Array)
                # und die entsprechenden Hintergrunddaten (X_val_shap_sample_df)
                shap.summary_plot(shap_values.values[:, :, i], X_val_shap_sample_df, max_display=MAX_SHAP_FEATURES, feature_names=features_in_model, show=False)
                plt.title(f"SHAP-Werte für Klasse: '{class_name}' - Neuronales Netz")
                plt.tight_layout()
                plt.show()
                
        elif isinstance(shap_values.values, list) and len(shap_values.values) > 1: # Falls es doch eine Liste von Arrays ist (unwahrscheinlich für DeepExplainer)
            print("INFO: SHAP-Werte sind Liste für Multiklasse. Aggregiere für globalen Bar-Plot.")
            shap_values_for_global_plot = np.mean(np.abs(np.array(shap_values.values)), axis=0) # Hier ist axis=0 korrekt
            
            shap.summary_plot(shap_values_for_global_plot, X_val_shap_sample_df, plot_type="bar", max_display=MAX_SHAP_FEATURES, show=False)
            plt.title("SHAP Feature Importance (Globaler Durchschnitt über alle Klassen) - Neuronales Netz")
            plt.tight_layout()
            plt.show()
            
            # Beeswarm Plots pro Klasse aus der Liste
            print("\n--> Erstelle SHAP Beeswarm Plots (Detailansicht) für jede Klasse...")
            if len(shap_values.values) == len(class_mapping):
                for i, class_name in enumerate(list(class_mapping.keys())):
                    print(f"    - Erstelle Beeswarm Plot für Klasse: '{class_name}' (Index {i}) - Neuronales Netz")
                    plt.figure(figsize=(12, 8))
                    shap.summary_plot(shap_values.values[i], X_val_shap_sample_df, max_display=MAX_SHAP_FEATURES, feature_names=features_in_model, show=False)
                    plt.title(f"SHAP-Werte für Klasse: '{class_name}' - Neuronales Netz")
                    plt.tight_layout()
                    plt.show()
            else:
                print(f"WARNUNG: Anzahl der SHAP-Werte-Listen ({len(shap_values.values)}) stimmt nicht mit der Anzahl der Klassen ({len(class_mapping)}) überein. Überspringe Beeswarm Plots pro Klasse.")
                
        else: # Dies wird ausgeführt, wenn shap_values.values weder 3D Array noch Liste > 1 Element ist
            print("INFO: SHAP-Werte sind kein Multiklassen-Format (oder nur eine Klasse). Erstelle einen einzelnen globalen Beeswarm/Bar-Plot.")
            # Für den globalen Plot bei nur einer Klasse, Beeswarm ist oft aussagekräftiger
            # Hier müssen wir shap_values.values übergeben (das sollte ein 2D-Array sein)
            shap.summary_plot(shap_values.values, X_val_shap_sample_df, max_display=MAX_SHAP_FEATURES, feature_names=features_in_model, show=False)
            plt.title("SHAP-Werte (Gesamt / Binäre Klassifikation) - Neuronales Netz")
            plt.tight_layout()
            plt.show()
            print("WARNUNG: Keine Beeswarm Plots pro Klasse möglich, da SHAP-Werte nicht im Multiklassen-Listenformat vorliegen.")
            
    else:
        print("Validierungsdaten sind leer, SHAP-Analyse wird übersprungen.")
except Exception as e:
    print(f"Fehler bei der Modell-Interpretation: {e}")
    import traceback
    traceback.print_exc()

print("\n[SUCCESS] Teil 3 abgeschlossen.")

# --- TEIL 4: SPEICHERN DER ERGEBNISSE ---
print("\n" + "="*80); print(" TEIL 4: SPEICHERN DER ERGEBNISSE"); print("=" * 80)
if y_test_pred is not None:
    test_results_df = X_test_raw.copy()
    test_results_df[label_column_ml] = y_test
    test_results_df['prediction'] = y_test_pred
    proba_df = pd.DataFrame(y_test_pred_proba, columns=[f'prediction_proba_{cls}' for cls in class_mapping.keys()], index=test_results_df.index)
    test_results_df = pd.concat([test_results_df, proba_df], axis=1)
    
    test_results_df.to_csv("test_data_final_classification_NEURAL_NET.csv", index=True)
    print(f"--> Testdaten mit Vorhersagen und Wahrscheinlichkeiten gespeichert: test_data_final_classification_NEURAL_NET.csv")
else:
    print("--> Testdaten sind leer, keine Speicherung.")

print("\nSkript vollständig abgeschlossen.")
```
