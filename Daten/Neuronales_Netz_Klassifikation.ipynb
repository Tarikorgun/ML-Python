```python
# --- TEIL 0: Bibliotheken importieren und Setup ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pastas as ps
import pyarrow.dataset as ds
import shap
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import gc
from statsmodels.tsa.seasonal import STL
from sklearn.model_selection import train_test_split # Bleibt für potenzielle andere Nutzungen, aber nicht für den Haupt-Split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils import class_weight

# +++ NEU: TensorFlow/Keras-Importe für das Neuronale Netz +++
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Loglevel reduzieren
ps.set_log_level("ERROR")
print("Starte Skript: SGI-Berechnung, Dekomposition und ML-Pipeline für DÜRRE-KLASSIFIKATION (mit Neuronalem Netz)")

# --- Definition der Features (unverändert) ---
STATIC_CATEGORICAL_FEATURES = ["main_landuse"]
STATIC_NUMERICAL_FEATURES = ["ground_elevation_m_asl", "topographic_slope_degrees"]
DYNAMIC_NUMERICAL_FEATURES_RAW = [
    "groundwater_level_m_asl", "precipitation_gpcc_mm_year-1",
    "potential_evapotranspiration_era5_mm_year-1", "air_temperature_°C", "snow_depth_m",
    "ndvi_ratio", "lai_low_vegetation_ratio", "lai_high_vegetation_ratio",
    "withdrawal_industrial_m3_year-1", "withdrawal_domestic_m3_year-1"
]
decomposition_plot_done = False
target_grow_id_for_decomposition_plot = None

# --- TEIL 1: Datenauswahl, SGI-Berechnung & Feature Collection (UNVERÄNDERT) ---

# KORRIGIERT: Feste, vollständige (absolute) Pfade zu Ihren Daten-Dateien.
path_attributes = "C:/MLmitPython/ML-Python/Daten/grow_attributes.parquet"
path_timeseries = "C:/MLmitPython/ML-Python/Daten/grow_timeseries.parquet"

try:
    attr_df_full = pd.read_parquet(path_attributes)
except FileNotFoundError:
    print(f"Fehler: Datei nicht gefunden unter dem Pfad '{path_attributes}'")
    print("Bitte stellen Sie sicher, dass der Pfad korrekt ist und die Datei dort existiert.")
    raise
except Exception as e:
    print(f"Fehler beim Laden von '{path_attributes}': {e}")
    raise

# Datentypen optimieren
for col in attr_df_full.select_dtypes(include=['float64']).columns:
    attr_df_full[col] = attr_df_full[col].astype(np.float32)
for col in attr_df_full.select_dtypes(include=['int64']).columns:
    if attr_df_full[col].min() >= np.iinfo(np.int32).min and attr_df_full[col].max() <= np.iinfo(np.int32).max:
        attr_df_full[col] = attr_df_full[col].astype(np.int32)

# Daten filtern
attr_filtered_df = attr_df_full[
    (attr_df_full["country"] == "CZE") &
    (attr_df_full["length_years"] >= 30) &
    (attr_df_full["gap_fraction"] == 0.0) &
    (pd.to_datetime(attr_df_full["ending_date"]) >= pd.Timestamp("2023-12-01"))
]
print(f"Gefundene Zeitreihen mit Kriterien für ML: {len(attr_filtered_df)}")
if attr_filtered_df.empty:
    raise ValueError("Keine passende Zeitreihe für CZ für ML gefunden.")

target_grow_id_for_decomposition_plot = attr_filtered_df.sort_values("length_years", ascending=False).iloc[0]["GROW_ID"]
print(f"Ziel-GROW_ID für den Dekompositionsplot: {target_grow_id_for_decomposition_plot}")

cz_ids_for_ml = attr_filtered_df["GROW_ID"].dropna().unique().tolist()
print(f"Anzahl eindeutiger GROW_IDs für ML-Datenbasis: {len(cz_ids_for_ml)}")

# Zeitreihen laden
dynamic_feature_columns_to_load = ["GROW_ID", "date"] + DYNAMIC_NUMERICAL_FEATURES_RAW
try:
    timeseries_dataset_raw = ds.dataset(path_timeseries, format="parquet")
    available_columns_in_parquet = timeseries_dataset_raw.schema.names
    columns_to_actually_load = [col for col in dynamic_feature_columns_to_load if col in available_columns_in_parquet]
    timeseries_table_raw = timeseries_dataset_raw.to_table(
        filter=ds.field("GROW_ID").isin(cz_ids_for_ml),
        columns=columns_to_actually_load
    )
    raw_ts_df_full = timeseries_table_raw.to_pandas(self_destruct=True, split_blocks=True, zero_copy_only=False)
except FileNotFoundError:
    print(f"Fehler: Datei nicht gefunden unter dem Pfad '{path_timeseries}'")
    print("Bitte stellen Sie sicher, dass der Pfad korrekt ist und die Datei dort existiert.")
    raise
except Exception as e:
    print(f"Fehler beim Laden von Zeitreihen: {e}")
    raise

# Zeitreihen vorverarbeiten
raw_ts_df_full["date"] = pd.to_datetime(raw_ts_df_full["date"])
if "precipitation_gpcc_mm_year-1" in raw_ts_df_full.columns:
    raw_ts_df_full["precipitation_gpcc_mm_month-1"] = (raw_ts_df_full["precipitation_gpcc_mm_year-1"] / 12).astype(np.float32)
if "potential_evapotranspiration_era5_mm_year-1" in raw_ts_df_full.columns:
    raw_ts_df_full["potential_evapotranspiration_era5_mm_month-1"] = (raw_ts_df_full["potential_evapotranspiration_era5_mm_year-1"] / 12).astype(np.float32)

cols_to_convert_float32 = [
    "groundwater_level_m_asl", "air_temperature_°C", "snow_depth_m", "ndvi_ratio",
    "lai_low_vegetation_ratio", "lai_high_vegetation_ratio",
    "withdrawal_industrial_m3_year-1", "withdrawal_domestic_m3_year-1"
]
for col in cols_to_convert_float32:
    if col in raw_ts_df_full.columns:
        raw_ts_df_full[col] = raw_ts_df_full[col].astype(np.float32)

grouped_ts_data = raw_ts_df_full.set_index('date').groupby('GROW_ID')
del raw_ts_df_full
gc.collect()

print(f"Beginne mit SGI-Berechnung, SGI-Dekomposition & Feature-Extraktion für {len(cz_ids_for_ml)} GROW_IDs...")

def process_grow_id_data(grow_id_data, current_grow_id):
    global decomposition_plot_done, target_grow_id_for_decomposition_plot
    gwl_series_raw = grow_id_data["groundwater_level_m_asl"].dropna()
    if gwl_series_raw.empty or len(gwl_series_raw.resample('MS').mean().dropna()) < 120:
        return None

    gwl_series_monthly = gwl_series_raw.resample('MS').mean().astype(np.float32)
    
    try:
        sgi_observed_monthly = ps.stats.sgi(gwl_series_monthly)
        if sgi_observed_monthly.empty or sgi_observed_monthly.isnull().all():
            return None
    except Exception:
        return None

    sgi_trend = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    sgi_seasonal = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    sgi_resid = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)

    sgi_for_stl = sgi_observed_monthly.dropna()
    if len(sgi_for_stl) >= 25:
        try:
            stl = STL(sgi_for_stl, period=12, robust=True)
            res_stl = stl.fit()
            sgi_trend = res_stl.trend.reindex(sgi_observed_monthly.index).astype(np.float32)
            sgi_seasonal = res_stl.seasonal.reindex(sgi_observed_monthly.index).astype(np.float32)
            sgi_resid = res_stl.resid.reindex(sgi_observed_monthly.index).astype(np.float32)
            
            if current_grow_id == target_grow_id_for_decomposition_plot and not decomposition_plot_done:
                print(f"\nErstelle Dekompositionsplot für SGI (GROW_ID: {current_grow_id})")
                fig_stl = res_stl.plot()
                fig_stl.suptitle(f"STL Dekomposition für SGI (GROW_ID: {current_grow_id})", fontsize=14)
                plt.tight_layout(rect=[0, 0, 1, 0.96])
                plt.show()
                decomposition_plot_done = True
        except Exception as e_stl:
            print(f"Fehler bei STL-Dekomposition für GROW_ID {current_grow_id}: {e_stl}.")

    monthly_features_dict = {'sgi_label': sgi_observed_monthly, 'sgi_trend': sgi_trend, 'sgi_seasonal': sgi_seasonal, 'sgi_resid': sgi_resid}
    
    dynamic_to_monthly_map = {
        "precipitation_gpcc_mm_month-1": "precipitation_monthly", "potential_evapotranspiration_era5_mm_month-1": "evapotranspiration_monthly",
        "air_temperature_°C": "air_temperature_monthly", "snow_depth_m": "snow_depth_monthly", "ndvi_ratio": "ndvi_monthly",
        "lai_low_vegetation_ratio": "lai_low_veg_monthly", "lai_high_vegetation_ratio": "lai_high_veg_monthly"
    }
    for col_orig, col_target in dynamic_to_monthly_map.items():
        if col_orig in grow_id_data.columns:
            monthly_features_dict[col_target] = grow_id_data[col_orig].dropna().resample('MS').mean().reindex(sgi_observed_monthly.index).astype(np.float32)
        else:
            monthly_features_dict[col_target] = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)

    yearly_to_monthly_ffill_map = {"withdrawal_industrial_m3_year-1": "withdrawal_industrial_monthly", "withdrawal_domestic_m3_year-1": "withdrawal_domestic_monthly"}
    for col_orig, col_target in yearly_to_monthly_ffill_map.items():
        if col_orig in grow_id_data.columns:
            series_yearly = grow_id_data[col_orig].dropna()
            if not series_yearly.index.is_unique: series_yearly = series_yearly[~series_yearly.index.duplicated(keep='first')]
            monthly_features_dict[col_target] = series_yearly.resample('MS').ffill().reindex(sgi_observed_monthly.index).astype(np.float32)
        else:
            monthly_features_dict[col_target] = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
            
    combined_monthly_df = pd.DataFrame(monthly_features_dict)
    combined_monthly_df.dropna(subset=['sgi_label'], inplace=True)
    return combined_monthly_df

processed_groups = []
for i, (grow_id, group_data) in enumerate(grouped_ts_data):
    if (i + 1) % 20 == 0: print(f"Verarbeite Gruppe {i+1}/{len(grouped_ts_data)}: {grow_id}")
    processed_df = process_grow_id_data(group_data, grow_id)
    if processed_df is not None and not processed_df.empty:
        processed_df['GROW_ID'] = grow_id
        processed_groups.append(processed_df)
del grouped_ts_data
gc.collect()

if not processed_groups: raise ValueError("Keine Daten für ML-Modell nach Verarbeitung.")
df_ml_input_dynamic = pd.concat(processed_groups)
del processed_groups
gc.collect()

static_features_to_select = ["GROW_ID"] + STATIC_NUMERICAL_FEATURES + STATIC_CATEGORICAL_FEATURES
static_features_to_select = [col for col in static_features_to_select if col in attr_df_full.columns]
attr_df_ml_static = attr_df_full[attr_df_full['GROW_ID'].isin(cz_ids_for_ml)][static_features_to_select].copy()
attr_df_ml_static.drop_duplicates(subset=['GROW_ID'], inplace=True, keep='first')

df_ml_input_dynamic.reset_index(inplace=True)
df_ml_input = pd.merge(df_ml_input_dynamic, attr_df_ml_static, on="GROW_ID", how="left").set_index('date')
del df_ml_input_dynamic, attr_df_ml_static, attr_df_full, attr_filtered_df
gc.collect()

print(f"Form des finalen ML-Input-DataFrames: {df_ml_input.shape}")
if df_ml_input.empty: raise ValueError("ML-Input-DataFrame ist leer.")


# --- TEIL 2: Machine-Learning-Pipeline für Klassifikation (UNVERÄNDERT) ---
print("Starte ML-Pipeline für Dürre-Klassifikation...")
df_ml_processed = df_ml_input.copy()
del df_ml_input
gc.collect()

# Lag-Features erstellen
for lag_col_base in ['sgi_label', 'sgi_trend', 'sgi_seasonal', 'sgi_resid']:
    if lag_col_base in df_ml_processed.columns:
        for lag_val in [1, 3, 6, 12]:
            target_col = f'{lag_col_base}_lag_{lag_val}M'
            df_ml_processed[target_col] = df_ml_processed.groupby('GROW_ID')[lag_col_base].shift(lag_val).astype(np.float32)

df_ml_processed['month_num'] = df_ml_processed.index.month.astype(np.int8)
df_ml_processed['year'] = df_ml_processed.index.year.astype(np.int32)

# Zukünftigen SGI-Wert für die Label-Erstellung berechnen
VORHERSAGE_HORIZONT_MONATE = 6
sgi_target_col_continuous = f'sgi_target_{VORHERSAGE_HORIZONT_MONATE}M_future'
df_ml_processed[sgi_target_col_continuous] = df_ml_processed.groupby('GROW_ID')['sgi_label'].shift(-VORHERSAGE_HORIZONT_MONATE).astype(np.float32)

def sgi_to_drought_class(z_score):
    if pd.isna(z_score): return None
    if z_score > -0.84: return 'Normal'
    if -1.50 <= z_score <= -0.84: return 'Mild_Moderate'
    if -2.00 <= z_score <= -1.51: return 'Severe'
    return 'Extreme'

df_ml_processed['drought_class_str'] = df_ml_processed[sgi_target_col_continuous].apply(sgi_to_drought_class)
class_mapping = {'Normal': 0, 'Mild_Moderate': 1, 'Severe': 2, 'Extreme': 3}
df_ml_processed['drought_class_target'] = df_ml_processed['drought_class_str'].map(class_mapping).astype('Int64')

print("\nVerteilung der Dürreklassen im Zieldatensatz:")
print(df_ml_processed['drought_class_target'].value_counts(normalize=True).sort_index())

# One-Hot Encoding
categorical_cols_to_encode = [col for col in STATIC_CATEGORICAL_FEATURES if col in df_ml_processed.columns] + ['month_num']
dummy_encoded_cols_list = []
for col_to_encode in categorical_cols_to_encode:
    print(f"One-Hot Encoding für Spalte: {col_to_encode}")
    dummies = pd.get_dummies(df_ml_processed[col_to_encode], prefix=col_to_encode, dummy_na=False, dtype=np.int8)
    df_ml_processed = pd.concat([df_ml_processed, dummies], axis=1)
    df_ml_processed.drop(columns=[col_to_encode], inplace=True)
    dummy_encoded_cols_list.extend(dummies.columns.tolist())

# Imputation
numerical_static_cols_for_imputation = [col for col in STATIC_NUMERICAL_FEATURES if col in df_ml_processed.columns]
if numerical_static_cols_for_imputation:
    imputer_numerical_static = SimpleImputer(strategy='median')
    df_ml_processed[numerical_static_cols_for_imputation] = imputer_numerical_static.fit_transform(df_ml_processed[numerical_static_cols_for_imputation]).astype(np.float32)

# Feature-Liste definieren
base_features_ml = [
    "precipitation_monthly", "evapotranspiration_monthly", "air_temperature_monthly",
    "snow_depth_monthly", "ndvi_monthly", "lai_low_veg_monthly", "lai_high_veg_monthly",
    "withdrawal_industrial_monthly", "withdrawal_domestic_monthly"
]
lag_features_ml = [col for col in df_ml_processed.columns if '_lag_' in col]
features_columns_ml = list(set(
    [col for col in base_features_ml if col in df_ml_processed.columns] +
    lag_features_ml +
    numerical_static_cols_for_imputation +
    dummy_encoded_cols_list
))
label_column_ml = 'drought_class_target'

# Finale Bereinigung
columns_to_check_for_dropna_smart = [
    label_column_ml,
    'sgi_label_lag_1M',
    'precipitation_monthly'
]
columns_to_check_for_dropna_smart = [col for col in columns_to_check_for_dropna_smart if col in df_ml_processed.columns]
print(f"\nForm des ML-DataFrames VOR dem finalen NaN-Drop: {df_ml_processed.shape}")
df_ml_processed.dropna(subset=columns_to_check_for_dropna_smart, inplace=True)
print(f"Form des ML-DataFrames NACH dem finalen NaN-Drop: {df_ml_processed.shape}")

if df_ml_processed.empty: raise ValueError("Keine Daten nach Bereinigung übrig.")

df_ml_processed[label_column_ml] = df_ml_processed[label_column_ml].astype(int)
X_all_features_raw = df_ml_processed[features_columns_ml].copy()
y_all_labels = df_ml_processed[label_column_ml].copy()
X_meta_for_split = df_ml_processed[['GROW_ID', 'year']].copy()
del df_ml_processed
gc.collect()

# ##########################################################################
# ### START DER ÄNDERUNGEN: AUSTAUSCH DES MODELLS GEGEN NEURONALES NETZ ###
# ##########################################################################

# --- TEIL 3: Train/Test Split, Skalierung, Training und Evaluation ---

# DYNAMISCHER, ZEITBASIERTER TRAIN/TEST SPLIT (UNVERÄNDERT)
temp_year_series = X_meta_for_split['year']
unique_years_ml = np.sort(temp_year_series.unique())
split_year_ml = int(np.percentile(unique_years_ml, 80))
print(f"\nDynamisches Split-Jahr für Training/Test: Daten vor {split_year_ml} für Training, ab {split_year_ml} für Test.")

train_mask_ml = temp_year_series < split_year_ml
test_mask_ml = temp_year_series >= split_year_ml
X_train_raw = X_all_features_raw[train_mask_ml]
X_test_raw = X_all_features_raw[test_mask_ml]
y_train = y_all_labels[train_mask_ml]
y_test = y_all_labels[test_mask_ml]

X_meta_test_ids = X_meta_for_split[test_mask_ml][['GROW_ID']]
del X_all_features_raw, y_all_labels, temp_year_series, X_meta_for_split
gc.collect()

print(f"Größe Roh-Trainingsdaten (X_train_raw): {X_train_raw.shape}")
print(f"Größe Roh-Testdaten (X_test_raw): {X_test_raw.shape}")
if X_train_raw.empty: raise ValueError("Trainingsdatensatz ist nach dem Split leer.")

# SKALIERUNG NUMERISCHER FEATURES (UNVERÄNDERT - ESSENTIELL FÜR NNs)
numerical_cols_to_scale = [col for col in X_train_raw.columns if col not in dummy_encoded_cols_list]
scaler = StandardScaler()
X_train = pd.DataFrame(index=X_train_raw.index)
X_test = pd.DataFrame(index=X_test_raw.index)

if numerical_cols_to_scale:
    print(f"Skaliere {len(numerical_cols_to_scale)} numerische Features...")
    # Wichtig: Imputer für NaN-Werte, die durch Lag-Operationen etc. entstanden sein könnten
    imputer_dynamic = SimpleImputer(strategy='mean')
    X_train_raw[numerical_cols_to_scale] = imputer_dynamic.fit_transform(X_train_raw[numerical_cols_to_scale])
    if not X_test_raw.empty:
        X_test_raw[numerical_cols_to_scale] = imputer_dynamic.transform(X_test_raw[numerical_cols_to_scale])

    X_train[numerical_cols_to_scale] = scaler.fit_transform(X_train_raw[numerical_cols_to_scale])
    if not X_test_raw.empty:
        X_test[numerical_cols_to_scale] = scaler.transform(X_test_raw[numerical_cols_to_scale])

# Hinzufügen der nicht-skalierten One-Hot-kodierten Features
for col in dummy_encoded_cols_list:
    if col in X_train_raw.columns: X_train[col] = X_train_raw[col].values
    if col in X_test_raw.columns: X_test[col] = X_test_raw[col].values

# Spaltenreihenfolge sicherstellen
X_train = X_train[features_columns_ml]
if not X_test.empty: X_test = X_test[features_columns_ml]
del X_train_raw, X_test_raw
gc.collect()

# --- MODELLTRAINING: Feed-Forward Neuronales Netz ---
print("\nStarte Modelltraining (Feed-Forward Neuronales Netz) mit Klassengewichtung...")

# Modellparameter definieren
n_features = X_train.shape[1]
n_classes = len(class_mapping)

# Definition der Modellarchitektur
ffn_model = Sequential([
    # Input Layer: Anzahl der Neuronen entspricht der Anzahl der Features
    Dense(128, activation='relu', input_shape=(n_features,)),
    Dropout(0.3), # Dropout zur Regularisierung
    # Hidden Layer 1
    Dense(64, activation='relu'),
    Dropout(0.3),
    # Hidden Layer 2
    Dense(32, activation='relu'),
    # Output Layer: Anzahl der Neuronen = Anzahl der Klassen, Softmax für Wahrscheinlichkeitsverteilung
    Dense(n_classes, activation='softmax')
])

# Modell kompilieren
ffn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', # Geeignet für Integer-Labels
    metrics=['accuracy']
)

print("\nModell-Architektur:")
ffn_model.summary()

# Klassengewichte berechnen, um die Imbalance im Datensatz auszugleichen
class_weights_values = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights_dict = dict(enumerate(class_weights_values))
print(f"\nBerechnete Klassengewichte: {class_weights_dict}")

# EarlyStopping-Callback, um Übertraining zu verhindern
early_stopping = EarlyStopping(
    monitor='val_loss', # Überwache den Fehler auf den Validierungsdaten
    patience=10,        # Anzahl der Epochen ohne Verbesserung, bevor das Training stoppt
    restore_best_weights=True # Stellt am Ende die besten gefundenen Gewichte wieder her
)

# Modell trainieren
history = ffn_model.fit(
    X_train,
    y_train,
    epochs=100, # Hohe Anzahl, da EarlyStopping das Training rechtzeitig beendet
    batch_size=64,
    validation_data=(X_test, y_test),
    class_weight=class_weights_dict,
    callbacks=[early_stopping],
    verbose=2 # Zeigt eine Zeile pro Epoche an
)

best_ffn_model = ffn_model # Da restore_best_weights=True ist

# --- EVALUATION ---
if not X_test.empty and not y_test.empty:
    # Vorhersagen für Testdaten generieren
    # .predict() gibt Wahrscheinlichkeiten für jede Klasse zurück
    y_pred_proba_ml = best_ffn_model.predict(X_test)
    # Die Klasse mit der höchsten Wahrscheinlichkeit als finale Vorhersage nehmen
    y_pred_ml = np.argmax(y_pred_proba_ml, axis=1)

    print("\n--- Klassifikations-Metriken auf Testdaten (Präzision, Recall, F1) ---")
    class_names = list(class_mapping.keys())
    print(classification_report(y_test, y_pred_ml, target_names=class_names))

    print("\n--- Klassifikationsmatrix (Confusion Matrix) ---")
    # Die Labels für die Matrix sind die Indizes der Klassen
    cm_labels = np.arange(n_classes)
    cm = confusion_matrix(y_test, y_pred_ml, labels=cm_labels)
    fig, ax = plt.subplots(figsize=(10, 8))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    disp.plot(cmap=plt.cm.Blues, ax=ax, xticks_rotation='vertical')
    ax.set_title(f'Klassifikationsmatrix für Dürrevorhersage ({VORHERSAGE_HORIZONT_MONATE}M Horizont)')
    plt.tight_layout()
    plt.show()

    # --- SHAP-Analyse (angepasst für Neuronale Netze) ---
    try:
        print("\nStarte SHAP-Analyse für das Neuronale Netz...")
        # DeepExplainer ist für Keras-Modelle optimiert.
        # Er benötigt ein "Hintergrund-Datenset", um die Erwartungswerte zu berechnen.
        # Ein Sample der Trainingsdaten ist hierfür üblich und performant.
        background_data_shap = X_train.sample(n=200, random_state=42)
        explainer_ffn = shap.DeepExplainer(best_ffn_model, background_data_shap)

        # SHAP-Werte für die Testdaten berechnen (kann einen Moment dauern)
        shap_values_ffn = explainer_ffn.shap_values(X_test)

        print("Zeige SHAP Feature Importance Plot...")
        # Für DeepExplainer ist der Output eine Liste von Arrays (eines pro Klasse)
        shap.summary_plot(shap_values_ffn, X_test, plot_type="bar", class_names=class_names, max_display=20)

        print("Zeige SHAP Beeswarm Plot (für alle Klassen)...")
        shap.summary_plot(shap_values_ffn, X_test, class_names=class_names, max_display=20)

    except Exception as e:
        print(f"Fehler bei SHAP-Berechnung/Darstellung: {e}")

# ##########################################################################
# ### ENDE DER ÄNDERUNGEN ###
# ##########################################################################

# --- TEIL 4: Speichern der Ergebnisse (UNVERÄNDERT) ---
print("\nSpeichere Trainings- und Testdaten mit Labels in CSV-Dateien...")
train_df_to_save = X_train.copy()
train_df_to_save[label_column_ml] = y_train
train_df_to_save.to_csv("train_data_classified_NN.csv", index=True)
print(f"Trainingsdaten gespeichert: train_data_classified_NN.csv mit Shape {train_df_to_save.shape}")

if not X_test.empty and not y_test.empty:
    test_df_to_save = X_test.copy()
    test_df_to_save[label_column_ml] = y_test
    test_df_to_save['prediction'] = y_pred_ml
    test_df_to_save.to_csv("test_data_classified_with_prediction_NN.csv", index=True)
    print(f"Testdaten gespeichert: test_data_classified_with_prediction_NN.csv mit Shape {test_df_to_save.shape}")
else:
    print("Testdaten sind leer, daher keine CSV-Datei für Testdaten erstellt.")

print("\nSkript vollständig abgeschlossen.")
```
