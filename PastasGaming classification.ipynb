```python
# --- TEIL 0: Bibliotheken importieren und Setup ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pastas as ps
import pyarrow.dataset as ds
import lightgbm as lgb
import shap
from sklearn.model_selection import GridSearchCV, train_test_split
# KLASSIFIKATION: Metriken für Klassifikation importieren
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler, LabelEncoder # LabelEncoder für die Zielvariable
from sklearn.impute import SimpleImputer
import gc # Garbage Collector
from statsmodels.tsa.seasonal import STL # Für Zeitreihen-Dekomposition

# Loglevel für Pastas und LightGBM reduzieren
ps.set_log_level("ERROR")
print("Starte Skript: SGI-Berechnung, SGI-Dekomposition und erweiterte ML-Pipeline (KLASSIFIKATION)")

# --- TEIL 1: Datenauswahl und SGI-Berechnung für das ML-Label & Feature Collection ---
try:
    attr_df_full = pd.read_parquet("Daten/grow_attributes.parquet")
except FileNotFoundError:
    print("Fehler: 'Daten/grow_attributes.parquet' nicht gefunden.")
    raise
except Exception as e:
    print(f"Fehler beim Laden von 'Daten/grow_attributes.parquet': {e}")
    raise

# Datentypen in attr_df frühzeitig optimieren
for col in attr_df_full.select_dtypes(include=['float64']).columns:
    attr_df_full[col] = attr_df_full[col].astype(np.float32)
for col in attr_df_full.select_dtypes(include=['int64']).columns:
    if attr_df_full[col].min() >= np.iinfo(np.int32).min and attr_df_full[col].max() <= np.iinfo(np.int32).max:
        attr_df_full[col] = attr_df_full[col].astype(np.int32)

attr_filtered_df = attr_df_full[
    (attr_df_full["country"] == "CZE") &
    (attr_df_full["length_years"] >= 30) &
    (attr_df_full["gap_fraction"] == 0.0) &
    (pd.to_datetime(attr_df_full["ending_date"]) >= pd.Timestamp("2023-12-01"))
]
print(f"Gefundene Zeitreihen mit Kriterien für ML: {len(attr_filtered_df)}")
if attr_filtered_df.empty:
    raise ValueError("Keine passende Zeitreihe für CZ für ML gefunden.")

cz_ids_for_ml = attr_filtered_df["GROW_ID"].dropna().unique().tolist()
print(f"Anzahl eindeutiger GROW_IDs für ML-Datenbasis: {len(cz_ids_for_ml)}")

dynamic_feature_columns_to_load = [
    "GROW_ID", "date", "groundwater_level_m_asl",
    "precipitation_gpcc_mm_year-1", "potential_evapotranspiration_era5_mm_year-1",
    "air_temperature_°C", "snow_depth_m", "ndvi_ratio",
    "lai_low_vegetation_ratio", "lai_high_vegetation_ratio",
    "withdrawal_industrial_m3_year-1", "withdrawal_domestic_m3_year-1"
]
try:
    timeseries_dataset_raw = ds.dataset("Daten/grow_timeseries.parquet", format="parquet")
    timeseries_table_raw = timeseries_dataset_raw.to_table(
        filter=ds.field("GROW_ID").isin(cz_ids_for_ml),
        columns=dynamic_feature_columns_to_load
    )
    raw_ts_df_full = timeseries_table_raw.to_pandas(self_destruct=True, split_blocks=True, zero_copy_only=False)
except FileNotFoundError:
    print("Fehler: 'Daten/grow_timeseries.parquet' nicht gefunden.")
    raise
except Exception as e:
    print(f"Fehler beim Laden von Zeitreihen: {e}")
    raise

raw_ts_df_full["date"] = pd.to_datetime(raw_ts_df_full["date"])
raw_ts_df_full["precipitation_gpcc_mm_month-1"] = (raw_ts_df_full["precipitation_gpcc_mm_year-1"] / 12).astype(np.float32)
raw_ts_df_full["potential_evapotranspiration_era5_mm_month-1"] = (raw_ts_df_full["potential_evapotranspiration_era5_mm_year-1"] / 12).astype(np.float32)
for col in ["groundwater_level_m_asl", "air_temperature_°C", "snow_depth_m", "ndvi_ratio",
            "lai_low_vegetation_ratio", "lai_high_vegetation_ratio",
            "withdrawal_industrial_m3_year-1", "withdrawal_domestic_m3_year-1"]:
    if col in raw_ts_df_full.columns:
        raw_ts_df_full[col] = raw_ts_df_full[col].astype(np.float32)

grouped_ts_data = raw_ts_df_full.set_index('date').groupby('GROW_ID')
del raw_ts_df_full
gc.collect()

all_ml_data_list = []
print(f"Beginne mit SGI-Berechnung, SGI-Dekomposition & Feature-Extraktion für {len(cz_ids_for_ml)} GROW_IDs...")

decomposition_plot_done = False # Flag für den Dekompositionsplot

def process_grow_id_data(grow_id_data, current_grow_id):
    global decomposition_plot_done

    gwl_series_raw = grow_id_data["groundwater_level_m_asl"].dropna()
    if gwl_series_raw.empty: return None
    gwl_series_monthly = gwl_series_raw.resample('MS').mean().astype(np.float32)
    
    if len(gwl_series_monthly.dropna()) < 120: # Genug Daten für robuste SGI-Berechnung
        return None
    
    try:
        sgi_observed_monthly = ps.stats.sgi(gwl_series_monthly)
        if sgi_observed_monthly.empty or sgi_observed_monthly.isnull().all():
            return None
    except Exception as e:
        return None

    sgi_trend = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    sgi_seasonal = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    sgi_resid = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)

    sgi_for_stl = sgi_observed_monthly.dropna()
    if len(sgi_for_stl) >= 25: 
        try:
            stl = STL(sgi_for_stl, period=12, robust=True)
            res_stl = stl.fit()
            
            sgi_trend = res_stl.trend.reindex(sgi_observed_monthly.index).astype(np.float32)
            sgi_seasonal = res_stl.seasonal.reindex(sgi_observed_monthly.index).astype(np.float32)
            sgi_resid = res_stl.resid.reindex(sgi_observed_monthly.index).astype(np.float32)

            # Plotting bleibt gleich, wird aber seltener aufgerufen, wenn viele IDs verarbeitet werden
            # if not decomposition_plot_done:
            #     print(f"\nErstelle Dekompositionsplot für SGI (GROW_ID: {current_grow_id})")
            #     try:
            #         fig_stl = res_stl.plot()
            #         fig_stl.suptitle(f"STL Dekomposition für SGI monatlich (GROW_ID: {current_grow_id})", fontsize=14)
            #         plt.tight_layout(rect=[0, 0, 1, 0.96])
            #         plt.show()
            #         decomposition_plot_done = True 
            #     except Exception as plot_e:
            #         print(f"  Fehler beim Erstellen des STL-Plots für {current_grow_id}: {plot_e}")
        except Exception as e:
            print(f"  Info: Fehler bei STL-Dekomposition des SGI für GROW_ID {current_grow_id}: {e}. Dekompositions-Features bleiben NaN.")

    monthly_features_dict = {
        'sgi_label': sgi_observed_monthly, 
        'sgi_trend': sgi_trend,
        'sgi_seasonal': sgi_seasonal,
        'sgi_resid': sgi_resid
    }
    
    for col_name_orig, col_name_target in [
        ("precipitation_gpcc_mm_month-1", "precipitation_monthly"),
        ("potential_evapotranspiration_era5_mm_month-1", "evapotranspiration_monthly"),
        ("air_temperature_°C", "air_temperature_monthly"),
        ("snow_depth_m", "snow_depth_monthly"),
        ("ndvi_ratio", "ndvi_monthly"),
        ("lai_low_vegetation_ratio", "lai_low_veg_monthly"),
        ("lai_high_vegetation_ratio", "lai_high_veg_monthly")
    ]:
        if col_name_orig in grow_id_data.columns:
            feat_series = grow_id_data[col_name_orig].dropna().resample('MS').mean()
            monthly_features_dict[col_name_target] = feat_series.reindex(sgi_observed_monthly.index).astype(np.float32)
        else:
            monthly_features_dict[col_name_target] = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    for col_name_orig, col_name_target in [
        ("withdrawal_industrial_m3_year-1", "withdrawal_industrial_monthly"),
        ("withdrawal_domestic_m3_year-1", "withdrawal_domestic_monthly")
    ]:
        if col_name_orig in grow_id_data.columns:
            feat_series_yearly = grow_id_data[col_name_orig].dropna()
            if not feat_series_yearly.index.is_unique:
                feat_series_yearly = feat_series_yearly[~feat_series_yearly.index.duplicated(keep='first')]
            feat_series_monthly_ffill = feat_series_yearly.resample('MS').ffill()
            monthly_features_dict[col_name_target] = feat_series_monthly_ffill.reindex(sgi_observed_monthly.index).astype(np.float32)
        else:
            monthly_features_dict[col_name_target] = pd.Series(index=sgi_observed_monthly.index, dtype=np.float32)
    
    combined_monthly_df = pd.DataFrame(monthly_features_dict)
    combined_monthly_df.dropna(subset=['sgi_label'], inplace=True)
    return combined_monthly_df

processed_groups = []
for i, (grow_id, group_data) in enumerate(grouped_ts_data):
    if (i + 1) % 100 == 0: # Seltener ausgeben, um Log nicht zu überfluten
        print(f"  Verarbeite Gruppe {i+1}/{len(grouped_ts_data)}: {grow_id}")
    processed_df = process_grow_id_data(group_data, grow_id)
    if processed_df is not None and not processed_df.empty:
        processed_df['GROW_ID'] = grow_id
        processed_groups.append(processed_df)
del grouped_ts_data
gc.collect()

# Einmaliger Dekompositionsplot für die erste verarbeitete ID mit genügend Daten
first_valid_df_for_plot = next((df for df in processed_groups if not df[['sgi_trend', 'sgi_seasonal', 'sgi_resid']].isnull().all().all()), None)
if first_valid_df_for_plot is not None:
    temp_sgi_for_plot = first_valid_df_for_plot['sgi_label'].dropna()
    if len(temp_sgi_for_plot) >=25:
        try:
            stl_plot = STL(temp_sgi_for_plot, period=12, robust=True).fit()
            fig_stl_plot = stl_plot.plot()
            fig_stl_plot.suptitle(f"Beispiel STL Dekomposition für SGI (GROW_ID: {first_valid_df_for_plot['GROW_ID'].iloc[0]})", fontsize=14)
            plt.tight_layout(rect=[0, 0, 1, 0.96])
            plt.show()
        except Exception as plot_e:
            print(f"Fehler beim Erstellen des finalen STL-Beispiel-Plots: {plot_e}")


print("SGI-Berechnung, SGI-Dekomposition & dynamische Feature-Extraktion für ML abgeschlossen.")
if not processed_groups:
    raise ValueError("Keine Daten für ML-Modell nach Verarbeitung.")

df_ml_input_dynamic = pd.concat(processed_groups)
del processed_groups
gc.collect()

static_features_to_select = [
    "GROW_ID", "ground_elevation_m_asl", "aquifer_type_class", "confinement",
    "soil_texture_0-30_cm_class", "topographic_slope_degrees", "main_landuse", "koeppen_geiger_class"
]
attr_df_ml_static = attr_df_full[attr_df_full['GROW_ID'].isin(cz_ids_for_ml)][static_features_to_select].copy()
attr_df_ml_static.drop_duplicates(subset=['GROW_ID'], inplace=True, keep='first')
for col in attr_df_ml_static.select_dtypes(include=['float64']).columns:
    attr_df_ml_static[col] = attr_df_ml_static[col].astype(np.float32)

df_ml_input_with_static = pd.merge(df_ml_input_dynamic.reset_index(), attr_df_ml_static, on="GROW_ID", how="left")
df_ml_input_with_static.set_index('date', inplace=True)
df_ml_input = df_ml_input_with_static

df_for_ids_in_split = df_ml_input[['GROW_ID']].copy() 
del df_ml_input_dynamic, attr_df_ml_static, attr_df_full, attr_filtered_df
gc.collect()

print(f"Form des ML-Input-DataFrames vor Feature Engineering: {df_ml_input.shape}")
if df_ml_input.empty:
    raise ValueError("ML-Input-DataFrame ist leer.")

# --- TEIL 2: Machine-Learning-Pipeline (KLASSIFIKATION)---
print("Starte erweiterte Machine-Learning-Pipeline mit SGI-Dekompositions-Features (KLASSIFIKATION)...")
df_ml_processed = df_ml_input.copy()
del df_ml_input
gc.collect()

for lag_col_base in ['sgi_label', 'sgi_trend', 'sgi_seasonal', 'sgi_resid']:
    if lag_col_base not in df_ml_processed.columns:
        print(f"Warnung: Spalte {lag_col_base} nicht in df_ml_processed gefunden. Überspringe Lag-Erstellung dafür.")
        continue
    for lag_val in [1, 3, 6, 12]:
        target_col = f'{lag_col_base}_lag_{lag_val}M'
        df_ml_processed[target_col] = df_ml_processed.groupby('GROW_ID')[lag_col_base].shift(lag_val).astype(np.float32)

df_ml_processed['month_num'] = df_ml_processed.index.month
df_ml_processed['year'] = df_ml_processed.index.year.astype(np.int32)

VORHERSAGE_HORIZONT_MONATE = 3
# KLASSIFIKATION: Zielvariable 'sgi_target_raw' erstellen (noch nicht klassifiziert)
df_ml_processed[f'sgi_target_raw_{VORHERSAGE_HORIZONT_MONATE}M_future'] = df_ml_processed.groupby('GROW_ID')['sgi_label'].shift(-VORHERSAGE_HORIZONT_MONATE).astype(np.float32)

# KLASSIFIKATION: Funktion zur Zuweisung der Dürreklassen
def sgi_to_drought_class(sgi_value):
    if pd.isna(sgi_value):
        return np.nan # Behalte NaNs, werden später entfernt
    if sgi_value < -2.00:
        return "Extreme"
    elif sgi_value <= -1.51: # -2.00 bis -1.51
        return "Severe"
    elif sgi_value <= -1.01: # -1.50 bis -1.01
        return "Moderate"
    elif sgi_value <= -0.84: # -1.00 bis -0.84
        return "Mild"
    else: # > -0.84
        return "Normal"

# KLASSIFIKATION: Erstelle die Klassen-Zielvariable
label_column_raw = f'sgi_target_raw_{VORHERSAGE_HORIZONT_MONATE}M_future'
label_column_class = f'sgi_class_target_{VORHERSAGE_HORIZONT_MONATE}M_future'
df_ml_processed[label_column_class] = df_ml_processed[label_column_raw].apply(sgi_to_drought_class)

# Überprüfe die Verteilung der Klassen
print("\nVerteilung der Dürreklassen in der Zielvariable:")
print(df_ml_processed[label_column_class].value_counts(dropna=False)) # dropna=False um auch NaNs zu sehen

categorical_cols_to_encode = ['aquifer_type_class', 'confinement', 'soil_texture_0-30_cm_class',
                                'main_landuse', 'koeppen_geiger_class', 'month_num']
for col in categorical_cols_to_encode:
    if col in df_ml_processed.columns:
        df_ml_processed[col] = df_ml_processed[col].fillna('Missing').astype(str)
    else:
        if col == 'month_num' : continue
        print(f"Warnung: Kategoriale Spalte {col} für One-Hot Encoding nicht in df_ml_processed gefunden.")

dummy_encoded_cols_list = []
for col_to_encode in categorical_cols_to_encode:
    if col_to_encode in df_ml_processed.columns:
        print(f"One-Hot Encoding für Spalte: {col_to_encode}")
        dummies = pd.get_dummies(df_ml_processed[col_to_encode], prefix=col_to_encode, dummy_na=False, dtype=np.int8) 
        df_ml_processed = pd.concat([df_ml_processed, dummies], axis=1)
        df_ml_processed.drop(columns=[col_to_encode], inplace=True)
        dummy_encoded_cols_list.extend(dummies.columns.tolist())
print(f"Form des ML-DataFrames nach One-Hot Encoding: {df_ml_processed.shape}")

numerical_static_cols = ['ground_elevation_m_asl', 'topographic_slope_degrees']
imputer_numerical = SimpleImputer(strategy='median')
for col in numerical_static_cols:
    if col in df_ml_processed.columns:
        df_ml_processed[col] = imputer_numerical.fit_transform(df_ml_processed[[col]]).astype(np.float32)

base_features_ml = [
    'precipitation_monthly', 'evapotranspiration_monthly', 'air_temperature_monthly',
    'snow_depth_monthly', 'ndvi_monthly', 'lai_low_veg_monthly', 'lai_high_veg_monthly',
    'withdrawal_industrial_monthly', 'withdrawal_domestic_monthly'
]
lag_features_sgi_label_ml = [f'sgi_label_lag_{lag}M' for lag in [1, 3, 6, 12]]
lag_features_sgi_trend_ml = [f'sgi_trend_lag_{lag}M' for lag in [1, 3, 6, 12] if f'sgi_trend_lag_{lag}M' in df_ml_processed.columns]
lag_features_sgi_seasonal_ml = [f'sgi_seasonal_lag_{lag}M' for lag in [1, 3, 6, 12] if f'sgi_seasonal_lag_{lag}M' in df_ml_processed.columns]
lag_features_sgi_resid_ml = [f'sgi_resid_lag_{lag}M' for lag in [1, 3, 6, 12] if f'sgi_resid_lag_{lag}M' in df_ml_processed.columns]
numerical_static_features_ml = [col for col in numerical_static_cols if col in df_ml_processed.columns]
features_columns_ml = (base_features_ml +
                       lag_features_sgi_label_ml +
                       lag_features_sgi_trend_ml +
                       lag_features_sgi_seasonal_ml +
                       lag_features_sgi_resid_ml +
                       numerical_static_features_ml +
                       dummy_encoded_cols_list)
features_columns_ml = [col for col in features_columns_ml if col in df_ml_processed.columns]
# KLASSIFIKATION: Zielspalte ist jetzt die Klassenspalte
label_column_ml = label_column_class 

# KLASSIFIKATION: Sicherstellen, dass NaNs in der Ziel-Klassenspalte und wichtigen Features entfernt werden
columns_to_check_for_dropna = [label_column_ml, 'sgi_label', 'precipitation_monthly',
                               'evapotranspiration_monthly', 'sgi_label_lag_1M',
                               ] + numerical_static_features_ml
if 'sgi_trend_lag_1M' in features_columns_ml: columns_to_check_for_dropna.append('sgi_trend_lag_1M')
if 'sgi_seasonal_lag_1M' in features_columns_ml: columns_to_check_for_dropna.append('sgi_seasonal_lag_1M')
columns_to_check_for_dropna = [col for col in columns_to_check_for_dropna if col in df_ml_processed.columns]

# Entferne Zeilen, wo die ZIELKLASSE NaN ist (entstanden durch shift oder weil sgi_target_raw NaN war)
df_ml_processed.dropna(subset=[label_column_ml], inplace=True)
# Entferne Zeilen mit NaNs in den für die Vorhersage kritischen Spalten
df_ml_processed.dropna(subset=columns_to_check_for_dropna, inplace=True)


print(f"Form des ML-DataFrames nach Feature Engineering und NaN-Drop (inkl. Zielklasse): {df_ml_processed.shape}")
if df_ml_processed.empty:
    raise ValueError("Keine Daten nach Feature Engineering und NaN-Bereinigung übrig.")

X_candidate_features = df_ml_processed[features_columns_ml].copy()
y_classification_str = df_ml_processed[label_column_ml].copy() # Noch als String-Labels

# KLASSIFIKATION: Label Encoding der Zielvariable (Strings zu Integers)
# Wichtig: Klassen müssen definiert und konsistent sein für LabelEncoder und spätere Plots
drought_classes_ordered = ["Extreme", "Severe", "Moderate", "Mild", "Normal"] # Für konsistente Reihenfolge
le = LabelEncoder()
le.fit(drought_classes_ordered) # Fitte den Encoder mit allen möglichen Klassen in gewünschter Reihenfolge
y = le.transform(y_classification_str) # Transformiere die tatsächlichen y-Werte

print("\nNumerische Kodierung der Klassen nach LabelEncoding:")
for i, class_name in enumerate(le.classes_):
    print(f"{class_name}: {i}")


X_meta_for_split = df_ml_processed[['GROW_ID', 'year']].copy()
del df_ml_processed
gc.collect()

numerical_features_to_scale = [
    col for col in X_candidate_features.columns
    if col not in dummy_encoded_cols_list 
]
numerical_features_to_scale = [col for col in numerical_features_to_scale if X_candidate_features[col].dtype in [np.float32, np.float64, np.int32, np.int64]]

X_final_for_model_parts = []
if numerical_features_to_scale:
    scaler = StandardScaler()
    scaled_values = scaler.fit_transform(X_candidate_features[numerical_features_to_scale])
    X_scaled_numerical_df = pd.DataFrame(scaled_values, columns=numerical_features_to_scale, index=X_candidate_features.index)
    X_final_for_model_parts.append(X_scaled_numerical_df)

X_dummy_df = X_candidate_features[dummy_encoded_cols_list]
X_final_for_model_parts.append(X_dummy_df)

X_final_for_model = pd.concat(X_final_for_model_parts, axis=1)
X_final_for_model = X_final_for_model[features_columns_ml]

del X_candidate_features, X_final_for_model_parts
if 'X_scaled_numerical_df' in locals(): del X_scaled_numerical_df
if 'X_dummy_df' in locals(): del X_dummy_df
gc.collect()

temp_year_series = X_meta_for_split['year']
unique_years_ml = np.sort(temp_year_series.unique())
X_train, X_test, y_train, y_test = None, None, None, None

# KLASSIFIKATION: Stratified Split, falls kein Zeit-Split möglich oder gewünscht
# Für Zeitreihen ist ein zeitlicher Split jedoch meist besser.
# Wenn `stratify=y` verwendet wird, muss `y` hier die Form vor dem Split haben.
y_for_stratify_split = y # y ist bereits ein numpy array der encoded labels

if len(unique_years_ml) <= 1:
    print("Warnung: Weniger als 2 Jahre Daten. Nutze zufälligen stratifizierten Split als Fallback.")
    # Um sicherzustellen, dass `train_test_split` die Indizes von X_final_for_model und y korrekt behandelt:
    indices = np.arange(len(X_final_for_model))
    try:
        # Versuche stratifizierten Split, wenn genügend Samples pro Klasse vorhanden sind
        train_indices, test_indices = train_test_split(indices, test_size=0.2, shuffle=True, random_state=42, stratify=y_for_stratify_split)
    except ValueError as e:
        print(f"Stratified Split fehlgeschlagen ({e}), nutze nicht-stratifizierten zufälligen Split.")
        train_indices, test_indices = train_test_split(indices, test_size=0.2, shuffle=True, random_state=42)
        
    X_train, X_test = X_final_for_model.iloc[train_indices], X_final_for_model.iloc[test_indices]
    y_train, y_test = y_for_stratify_split[train_indices], y_for_stratify_split[test_indices] # y ist hier das numerisch kodierte Array
    X_meta_train = X_meta_for_split.iloc[train_indices]

else:
    split_year_index_ml = int(len(unique_years_ml) * 0.8)
    if split_year_index_ml >= len(unique_years_ml): split_year_index_ml = len(unique_years_ml) - 1
    split_year_ml = unique_years_ml[split_year_index_ml if split_year_index_ml > 0 else 0]
    print(f"Split-Jahr für Training/Test: Daten vor {split_year_ml} für Training, ab {split_year_ml} für Test.")
    
    train_mask_ml = temp_year_series < split_year_ml
    test_mask_ml = temp_year_series >= split_year_ml
    
    X_train = X_final_for_model[train_mask_ml]
    X_test = X_final_for_model[test_mask_ml]
    y_train = y_for_stratify_split[train_mask_ml.values] # .values um sicherzustellen, dass boolesche Indizierung korrekt funktioniert
    y_test = y_for_stratify_split[test_mask_ml.values]
    X_meta_train = X_meta_for_split[train_mask_ml]

del X_final_for_model, y_for_stratify_split, temp_year_series, X_meta_for_split # y (Original) wird nicht mehr benötigt
gc.collect()

print(f"Größe Trainingsdaten (X_train): {X_train.shape}, (y_train): {y_train.shape}")
print(f"Größe Testdaten (X_test): {X_test.shape}, (y_test): {y_test.shape}")

if X_train.empty or X_test.empty or y_train.size == 0 or y_test.size == 0 :
    raise ValueError("Trainings- oder Testdatensatz (Features oder Label) nach Split leer.")

unique_grow_ids_in_train = X_meta_train['GROW_ID'].nunique()
print(f"Modell wird mit {X_train.shape[0]} Datenpunkten von {unique_grow_ids_in_train} eindeutigen Zeitreihen (GROW_IDs) trainiert.")
del X_meta_train
gc.collect()

print("\nKlassenverteilung im Trainingsdatensatz:")
unique_train, counts_train = np.unique(y_train, return_counts=True)
print(dict(zip(le.inverse_transform(unique_train), counts_train)))

print("\nKlassenverteilung im Testdatensatz:")
unique_test, counts_test = np.unique(y_test, return_counts=True)
print(dict(zip(le.inverse_transform(unique_test), counts_test)))


# KLASSIFIKATION: LGBMClassifier verwenden
# Bei stark unbalancierten Klassen: class_weight='balanced' oder manuelle Gewichte
lgbm_model_clf = lgb.LGBMClassifier(random_state=42, verbosity=-1) # class_weight='balanced'
param_grid_lgbm_clf = {'n_estimators': [100, 200], # Reduzierte Parameter für schnellere Demo
                       'learning_rate': [0.05, 0.1], 
                       'num_leaves': [31, 50]}
# KLASSIFIKATION: Scoring-Metrik für Klassifikation (z.B. f1_weighted für unbalancierte Klassen)
grid_search_lgbm_clf = GridSearchCV(estimator=lgbm_model_clf, param_grid=param_grid_lgbm_clf, cv=3, # Reduzierte CV für schnellere Demo
                                scoring='f1_weighted', verbose=1, n_jobs=-1) # f1_weighted ist gut für Imbalance

try:
    grid_search_lgbm_clf.fit(X_train, y_train)
    best_lgbm_clf_model = grid_search_lgbm_clf.best_estimator_
    print(f"Beste Parameter für LightGBM Classifier: {grid_search_lgbm_clf.best_params_}")

    y_pred_clf = best_lgbm_clf_model.predict(X_test)
    y_pred_proba_clf = best_lgbm_clf_model.predict_proba(X_test) # Für SHAP und andere Metriken

    # KLASSIFIKATION: Metriken ausgeben
    accuracy_clf = accuracy_score(y_test, y_pred_clf)
    print(f"\nAccuracy auf Testdaten (Vorhersagehorizont {VORHERSAGE_HORIZONT_MONATE} Monate): {accuracy_clf:.4f}")
    
    print("\nClassification Report auf Testdaten:")
    # Stelle sicher, dass target_names die String-Labels in der korrekten, vom LabelEncoder verwendeten Reihenfolge sind
    print(classification_report(y_test, y_pred_clf, target_names=le.classes_))

    print("\nConfusion Matrix auf Testdaten:")
    cm = confusion_matrix(y_test, y_pred_clf, labels=le.transform(le.classes_)) # stelle sicher, dass Labels die numerischen Werte sind
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix (SGI Klassifikation, {VORHERSAGE_HORIZONT_MONATE}M Vorh.)")
    plt.show()

except ValueError as e_fit:
    print(f"Fehler während des Modelltrainings oder der Evaluation: {e_fit}")
    print("Dies kann passieren, wenn eine Klasse im Trainings- oder Testset nach dem Split fehlt oder zu wenige Samples hat.")
    best_lgbm_clf_model = None # Setze auf None, um spätere Fehler zu vermeiden

if best_lgbm_clf_model: # Nur wenn das Modell erfolgreich trainiert wurde
    try:
        # Für Klassifikation: shap_values gibt eine Liste von Arrays (eines pro Klasse)
        explainer_lgbm_clf = shap.TreeExplainer(best_lgbm_clf_model)
        shap_values_lgbm_clf = explainer_lgbm_clf.shap_values(X_test) 
        # shap_values_lgbm_clf ist eine Liste von Arrays (für jede Klasse ein Array)
        # Für den summary_plot können wir die SHAP-Werte für eine bestimmte Klasse oder den Durchschnitt verwenden.
        # Oder wir plotten für jede Klasse einzeln, wenn gewünscht.
        # Hier verwenden wir den Durchschnitt der absoluten SHAP-Werte über alle Klassen für die Bar-Plot-Wichtigkeit
        # Oder wir nehmen die SHAP-Werte für eine spezifische, wichtige Klasse.
        # Für den Beeswarm-Plot nimmt man oft die Werte für eine Klasse.

        # Bar Plot (mittlere absolute SHAP Werte über alle Klassen)
        # Um einen einzelnen Wert pro Feature zu bekommen, mitteln wir die Absolutwerte der SHAP Values über alle Klassen
        # shap_values_for_bar = np.abs(shap_values_lgbm_clf).mean(0) # Nicht direkt unterstützt von shap.summary_plot
        
        print("\nErstelle SHAP Summary Plots (kann bei vielen Daten dauern)...")
        # Bar Plot - aggregiert über alle Klassen
        # shap.summary_plot(shap_values_lgbm_clf, X_test, plot_type="bar", class_names=le.classes_, show=False, max_display=30)
        # Alternative für Bar-Plot: Wichtigkeit für jede Klasse
        # for i, class_name in enumerate(le.classes_):
        # plt.figure(figsize=(12, int(max(5, 0.2 * len(X_test.columns))) +2 ))
        # shap.summary_plot(shap_values_lgbm_clf[i], X_test, plot_type="bar", show=False, max_display=30)
        # plt.title(f"ML: Agg. SHAP Feature Importance für Klasse '{class_name}' ({VORHERSAGE_HORIZONT_MONATE}M Vorh.)")
        # plt.tight_layout()
        # plt.show()
        
        # Da shap.summary_plot(plot_type="bar") mit Listen von Arrays (multi-class) manchmal Probleme macht
        # oder nicht die gewünschte Aggregation liefert, hier ein Beeswarm für jede Klasse:
        for i, class_name in enumerate(le.classes_):
            if isinstance(shap_values_lgbm_clf, list) and i < len(shap_values_lgbm_clf):
                plt.figure(figsize=(12, int(max(5, 0.2 * len(X_test.columns))) +2 ))
                shap.summary_plot(shap_values_lgbm_clf[i], X_test, show=False, max_display=30)
                plt.title(f"ML: SHAP Beeswarm Plot für Klasse '{class_name}' ({VORHERSAGE_HORIZONT_MONATE}M Vorh.)")
                plt.tight_layout()
                plt.show()
            else: # Für binäre Klassifikation ist shap_values_lgbm_clf oft nur ein Array
                if i == 1: # Bei Binärklassifikation oft nur für die positive Klasse relevant
                    plt.figure(figsize=(12, int(max(5, 0.2 * len(X_test.columns))) +2 ))
                    shap.summary_plot(shap_values_lgbm_clf, X_test, show=False, max_display=30) # Hier direkt shap_values_lgbm_clf
                    plt.title(f"ML: SHAP Beeswarm Plot ({VORHERSAGE_HORIZONT_MONATE}M Vorh.)")
                    plt.tight_layout()
                    plt.show()
                break # Nur einmal plotten, wenn nicht multi-class Output von SHAP


    except Exception as e:
        print(f"Fehler bei SHAP-Berechnung/Darstellung für Klassifikationsmodell: {e}")
        if isinstance(e, MemoryError):
            print("  -> SHAP-Berechnung erfordert möglicherweise zu viel Speicher.")

# Kein direkter Plot von Vorhersage vs. Realität bei Klassifikation, da die y-Achse Klassen wären.
# Die Confusion Matrix und der Classification Report sind hier die primären Evaluationswerkzeuge.

print(f"Erweiterte Machine-Learning-Pipeline (KLASSIFIKATION, SGI-Dekomposition, One-Hot, {VORHERSAGE_HORIZONT_MONATE}M Vorh.) abgeschlossen.")
```
